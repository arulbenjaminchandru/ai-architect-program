## 🧠 What is RAG (in one minute)?

An LLM only knows what it was trained on — not *your* company PDFs. **RAG (Retrieval-Augmented Generation)** fixes this: before answering, the app **retrieves** the most relevant chunks from your documents and **feeds them to Claude** as context. Claude then answers using *your* data.

> Think of it like an open-book exam. Claude is smart, but RAG hands it the right page of *your* textbook before it answers.

```
  ┌──────────┐   chunk+index   ┌────────────┐
  │ Your docs│ ───────────────►│ Vector DB  │
  └──────────┘                 └─────┬──────┘
                                     │ retrieve top matches
   User question ───────────────────►│
                                     ▼
                            ┌─────────────────┐
                            │  Claude (LLM)   │──► grounded answer
                            └─────────────────┘
```

## 📋 Reference: ingest.py (what Claude should produce)
If you want to check Claude's output against a known-good version:

```python
import os, glob
import chromadb
from pypdf import PdfReader

def load_text(path):                       # read a .txt or .pdf file into one string
    if path.endswith(".pdf"):
        reader = PdfReader(path)
        return "\n".join(page.extract_text() or "" for page in reader.pages)
    with open(path, encoding="utf-8") as f:
        return f.read()

def chunk(text, size=500):                 # split into ~500-char pieces
    return [text[i:i+size] for i in range(0, len(text), size)]

client = chromadb.PersistentClient(path="./chroma_db")
collection = client.get_or_create_collection("docs")

count = 0
for path in glob.glob("documents/*"):      # every file in documents/
    for i, piece in enumerate(chunk(load_text(path))):
        collection.add(documents=[piece], ids=[f"{os.path.basename(path)}-{i}"])
        count += 1
print(f"✅ Indexed {count} chunks")
```

## 📋 Reference: rag_chat.py (what Claude should produce)

```python
import os
import anthropic
import chromadb

client = anthropic.Anthropic()            # reads ANTHROPIC_API_KEY from the env
collection = chromadb.PersistentClient(path="./chroma_db").get_collection("docs")

print("Ask a question (or 'quit'):")
while True:
    question = input("> ")
    if question.lower() == "quit":
        break

    # 1) RETRIEVE: find the 3 most relevant chunks from your documents
    results = collection.query(query_texts=[question], n_results=3)
    context = "\n\n".join(results["documents"][0])

    # 2) AUGMENT: build a prompt that includes that context
    prompt = (
        f"Answer using ONLY the context below.\n\n"
        f"CONTEXT:\n{context}\n\nQUESTION: {question}"
    )

    # 3) GENERATE: let Claude answer from your data
    response = client.messages.create(
        model="claude-haiku-4-5-20251001",
        max_tokens=500,
        messages=[{"role": "user", "content": prompt}]
    )
    print("\n" + response.content[0].text + "\n")
```

# 🧩 Part B (Bonus Lab) — Build an **LLM Wiki** (the "Beyond-RAG" pattern)

You just built a RAG chatbot. Now build its successor: **Karpathy's LLM Wiki** — a knowledge base where an LLM *compiles* your raw sources into structured markdown once, then reads from that compiled layer forever.

> In **April 2026** Andrej Karpathy posted a GitHub Gist describing this pattern. It hit **16M+ views** because it solves the problem every knowledge base has: maintenance. The LLM does the filing, cross-referencing, and summarising — you just curate what goes in.

We'll build one over **your own course notebooks** (these `.ipynb` lab files) using **Claude Code**, the same tool from Part A.

## 🧠 RAG vs. LLM Wiki — what actually changes

In Part A, every question triggered a *fresh retrieval* from static chunks. The model rediscovered your docs from scratch each time. The LLM Wiki flips this: **compile once, read forever.** Knowledge **compounds** — each new source improves every future answer.

| Dimension | RAG (Part A) | LLM Wiki (Part B) |
|---|---|---|
| **State** | Stateless — each query independent | **Stateful** — knowledge compounds |
| **Infrastructure** | Vector DB + embeddings + retrieval | A **folder of `.md` files** |
| **Cross-references** | Re-discovered per query | **Pre-built** by the LLM, always there |
| **Token cost / query** | High (retrieve + re-rank + generate) | Low (read index + a few pages) |
| **Traceability** | Chunk-level (often lossy) | **Source-level** — every claim links to `raw/` |
| **Contradictions** | Undetected | **Flagged** during *lint* |
| **Sweet spot** | Millions of docs | **< ~100–200 docs / < ~80k tokens** |

> **Rule of thumb:** under ~100 docs, skip the vector DB — the LLM just reads the wiki directly. That's the **70x-more-efficient** claim: no embedding pipeline, no re-ranking, just markdown.

## 🏗️ The three-layer architecture

The whole pattern is a **compiler analogy**: `raw/` is source code → the LLM is the compiler → `wiki/` is the compiled executable → *lint* is the test suite → *query* is runtime.

```
my-wiki/
├── raw/                 # LAYER 1 — immutable sources (you add these, LLM never edits)
│   ├── notebooks/       #   e.g. your Day_01.ipynb … Day_14.ipynb
│   ├── articles/
│   ├── papers/
│   └── images/
├── wiki/                # LAYER 2 — LLM-OWNED markdown (LLM writes; you mostly read)
│   ├── index.md         #   master catalog — LLM reads this FIRST to navigate
│   ├── log.md           #   append-only history of every operation
│   ├── concepts/        #   concept pages  (rag.md, vector-db.md, claude-code.md …)
│   ├── entities/        #   entity pages   (anthropic.md, chromadb.md …)
│   ├── sources/         #   one summary per ingested source
│   └── comparisons/     #   rag-vs-llm-wiki.md …
├── outputs/             # dated reports, lint results, slides
└── CLAUDE.md            # LAYER 3 — the SCHEMA that governs how the LLM maintains the wiki
```

1. **Raw (immutable):** your curated sources. The verification baseline — every wiki claim traces back here.
2. **Wiki (LLM-owned):** entity pages, concept pages, source summaries, an index, and a log — all markdown.
3. **Schema (`CLAUDE.md`):** the config that turns a generic agent into a *disciplined wiki maintainer*. **This is the most important file.**

## 🔁 The three operations: Ingest → Query → Lint

Everything you do with the wiki is one of these three verbs.

| Operation | You do | The LLM does |
|---|---|---|
| **1. Ingest** | Drop a source in `raw/`, say *"ingest it"* | Reads it, discusses takeaways, writes a `sources/` summary, updates/creates concept & entity pages, updates `index.md`, appends to `log.md` |
| **2. Query** | Ask a question | Reads `index.md`, opens only the relevant pages, answers **with `[[wikilink]]` citations**, optionally files the answer back as a new page |
| **3. Lint** | Say *"lint the wiki"* (periodically) | Scans for contradictions, orphan pages, missing concepts, stale claims, and research gaps → saves a report to `outputs/` |

> **Two key principles:** **Compile-first** (don't just answer — write conclusions *into* wiki pages) and **Writeback is mandatory** (every valuable finding goes back to the wiki). That's why knowledge compounds.

## 🧾 Step 18 — Write `CLAUDE.md` (the schema — the most important file)
This file turns a generic agent into a *disciplined* one. Without it, output drifts and pages are inconsistent. With it, the LLM behaves like a reliable knowledge worker.

### 📋 Reference: `CLAUDE.md` (a known-good schema)
```markdown
# LLM Wiki: <Your Topic>

## Project Structure
- `raw/`  — Immutable source documents. NEVER modify files here.
- `wiki/` — LLM-generated & maintained markdown pages.
- `wiki/index.md` — Master content catalog. Update on EVERY operation.
- `wiki/log.md`   — Append-only operation log.
- `outputs/` — Generated reports, presentations, lint results.

## Page Types & Conventions
Every wiki page MUST start with YAML frontmatter:

---
title: Page Title
type: concept | entity | source-summary | comparison
sources:
  - raw/notebooks/Day_14.ipynb
related:
  - "[[related-concept]]"
created: YYYY-MM-DD
updated: YYYY-MM-DD
confidence: high | medium | low
---

- Filenames: kebab-case matching the concept (e.g. attention-mechanism.md)
- Internal links: use [[wikilinks]] everywhere
- Always link claims back to their raw/ source path

## Workflows
### Ingest
1. Read the source in raw/
2. Discuss key takeaways with me
3. Create wiki/sources/<source-name>.md summary
4. Update or create concept/entity pages as needed
5. Update wiki/index.md
6. Append to wiki/log.md (date, source, pages touched, notable findings)

### Query
1. Read wiki/index.md to find relevant pages
2. Read only those pages; synthesize an answer
3. Cite with [[wikilinks]] back to sources
4. If the answer is novel & valuable, offer to save it as a new wiki page

### Lint
1. Scan all wiki pages for contradictions
2. Find orphan pages (no incoming [[wikilinks]])
3. Flag missing concepts (referenced but no page)
4. Flag stale/low-confidence claims superseded by newer sources
5. Save results to outputs/lint-YYYY-MM-DD.md
```
> Customise it for your domain — e.g. add a `prerequisites:` field for course notes, or a `difficulty:` tag.

## ✅ Step 25 — Test your wiki
1. **Compounding:** ingest a second source on the same topic, then re-ask an earlier question — the answer should now reference *both* sources and any new cross-links.
2. **Traceability:** open any wiki page and confirm its `sources:` frontmatter points to a real file in `raw/`.
3. **Index-first navigation:** ask something niche and watch Claude open `index.md` before any concept page (low token cost, not a brute-force scan).
4. **Lint catches drift:** deliberately add a contradicting note, ingest it, then lint — Claude should flag the conflict.

## 🔭 When to use which (RAG vs. Wiki)
Keep **both** tools from this lab in your kit:

- **Use the LLM Wiki** when you have **fewer than ~100–200 sources**, want knowledge to **compound**, care about **source-level traceability**, and want **zero infrastructure** (just a folder + Claude).
- **Use RAG (Part A)** for **large, static corpora** (millions of docs), frequently-changing collections, **sub-second latency at scale**, or multi-team access control.

> They're not rivals. Compile your *working* knowledge into a wiki; reach for RAG when the corpus outgrows what an LLM can read directly.

# 🔌 Part C (Bonus Lab) — Run **Claude Code on Gemini** with LiteLLM

So far Claude Code talked to Anthropic's own models. But Claude Code speaks the **Anthropic Messages API**, and so does a **LiteLLM proxy** — so you can put LiteLLM in the middle and route every request to **Google Gemini** (or OpenAI, DeepSeek, Bedrock…) while keeping the exact same Claude Code workflow from Parts A & B.

```
  ┌─────────────┐   Anthropic format   ┌────────────┐   Gemini format   ┌──────────┐
  │ Claude Code │ ───────────────────► │  LiteLLM   │ ────────────────► │  Gemini  │
  │   (CLI)     │ ◄─────────────────── │  proxy     │ ◄──────────────── │  (Google)│
  └─────────────┘   translated back    └────────────┘                   └──────────┘
         :4000 (localhost)
```

> **Why bother?** One CLI, any model. LiteLLM handles translation, fallbacks, and cost tracking — and it's how you use Claude Code when you have Gemini/OpenAI keys instead of an Anthropic one.

## ✅ Prerequisites
- **Claude Code** installed (from Part A)
- **Python 3.10+** and `pip`
- A **Google AI Studio API key** — get one free at https://aistudio.google.com/apikey

> ⚠️ **Security note (read this):** Anthropic flagged that LiteLLM PyPI versions **1.82.7** and **1.82.8** shipped credential-stealing malware. **Do not install those two versions.** Use a newer clean release, and if you ever installed them, rotate your API keys.

## 📦 Step 26 — Install LiteLLM (with proxy support)
```bash
pip install 'litellm[proxy]'
```

> The `[proxy]` extra pulls in the server bits you need to run the local gateway. (If you use **uv**, the official equivalent is `uv tool install 'litellm[proxy]'`.)

## 🧾 Step 27 — Create the config file
In your project folder, create **`litellm_config.yaml`**. This maps a friendly model name to a real Gemini model:

```yaml
model_list:
  - model_name: gemini-2.5-flash
    litellm_params:
      model: gemini/gemini-2.5-flash
      api_key: YOUR_GOOGLE_API_KEY
```

Replace `YOUR_GOOGLE_API_KEY` with your key from **Google AI Studio**.

> 🔐 **Better practice (don't hard-code keys):** put the key in your environment and reference it instead, so the secret never lives in the file:
> ```yaml
> model_list:
>   - model_name: gemini-2.5-flash
>     litellm_params:
>       model: gemini/gemini-2.5-flash
>       api_key: os.environ/GEMINI_API_KEY
> ```
> then `export GEMINI_API_KEY="your-key"` before starting the proxy.

## 🚀 Step 28 — Start the proxy
```bash
litellm --config litellm_config.yaml --port 4000
```

You should see LiteLLM start and report that it's **listening on port 4000** (`RUNNING on http://0.0.0.0:4000`). **Leave this terminal running** — it's your gateway.

> 💡 Open a **second terminal** for the next steps so the proxy keeps running in the first.

## 🔎 Step 29 — (Optional) Verify the proxy works
Before wiring up Claude Code, confirm the gateway can reach Gemini. In a new terminal:

```bash
curl -X POST http://localhost:4000/v1/messages \
  -H "Content-Type: application/json" \
  -d '{
    "model": "gemini-2.5-flash",
    "max_tokens": 100,
    "messages": [{"role": "user", "content": "Say hello in one short sentence."}]
  }'
```

A JSON reply with text means LiteLLM is translating Anthropic ↔ Gemini correctly. If this fails, fix it here before moving on.

## ⚙️ Step 30 — Configure Claude Code
Create a **`.env`** file in your project folder that points Claude Code at the proxy instead of Anthropic:

```bash
ANTHROPIC_BASE_URL=http://localhost:4000
ANTHROPIC_API_KEY=DUMMY_KEY
ANTHROPIC_AUTH_TOKEN=DUMMY_KEY
ANTHROPIC_MODEL=gemini-2.5-flash
```

What each line does:

| Variable | Purpose |
|---|---|
| `ANTHROPIC_BASE_URL` | Sends Claude Code's requests to **your local proxy**, not api.anthropic.com |
| `ANTHROPIC_API_KEY` / `ANTHROPIC_AUTH_TOKEN` | The token Claude Code sends to the proxy. It's a **dummy** because your real Gemini key lives in LiteLLM. *(If you set a `LITELLM_MASTER_KEY`, use that value here instead.)* |
| `ANTHROPIC_MODEL` | The default model name — must match the `model_name` in your config (`gemini-2.5-flash`) |

## ▶️ Step 31 — Load the env and launch Claude Code
Load the variables into your shell, then start Claude Code:

```bash
set -a && source .env && set +a
claude
```

`set -a` auto-exports every variable defined while sourcing `.env`; `set +a` turns that off again. **That's it** — Claude Code now connects to LiteLLM, which routes everything to **Gemini**.

> 🪟 **Windows (PowerShell)** doesn't have `source`. Set them inline instead:
> ```powershell
> $env:ANTHROPIC_BASE_URL="http://localhost:4000"
> $env:ANTHROPIC_API_KEY="DUMMY_KEY"
> $env:ANTHROPIC_AUTH_TOKEN="DUMMY_KEY"
> $env:ANTHROPIC_MODEL="gemini-2.5-flash"
> claude
> ```

## ✅ Step 32 — Confirm it's really using Gemini
1. In Claude Code, ask: `Which model are you and who made you?` — a Gemini-backed answer confirms the routing.
2. Watch the **LiteLLM terminal**: each prompt logs a request being routed to `gemini/gemini-2.5-flash`.
3. Try a real task (e.g. *"create hello.py that prints the first 10 Fibonacci numbers"*) — the full Claude Code edit/diff loop now runs on Gemini.

> 🎓 **Tip for learners:** add more entries under `model_list` (e.g. an OpenAI or a second Gemini model) and switch between them with `claude --model <name>` — one proxy, many models.

## 🛠️ Troubleshooting (Part C)
| Problem | Cause | Fix |
|---|---|---|
| `Connection refused` on :4000 | Proxy not running / wrong port | Keep the Step 28 terminal open; confirm it says `RUNNING on ...:4000` |
| `401 / invalid api key` | Gemini key wrong or not loaded | Check `YOUR_GOOGLE_API_KEY` (or `GEMINI_API_KEY` export) is a valid AI Studio key |
| Claude Code still hits Anthropic | `.env` not loaded | Re-run `set -a && source .env && set +a` in the **same** shell before `claude` |
| `model not found` | Name mismatch | `ANTHROPIC_MODEL` must equal the `model_name` in `litellm_config.yaml` |
| Weird security warning | Bad LiteLLM version | Avoid 1.82.7 / 1.82.8; upgrade with `pip install -U 'litellm[proxy]'` and rotate keys |

## 🎁 Wrap-up
**Built:** a local **LiteLLM proxy** that lets **Claude Code run on Google Gemini** — same CLI, same workflow, different engine.

**Learned:** Claude Code talks Anthropic-format; a proxy can translate that to *any* provider. Point `ANTHROPIC_BASE_URL` at the proxy, give it a dummy token, set `ANTHROPIC_MODEL`, and you're routing through Gemini.
