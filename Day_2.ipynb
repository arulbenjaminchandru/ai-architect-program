# 💰 Claude Pricing: API vs Claude.ai Subscriptions

Claude (by Anthropic) offers **two main pricing models**:
1. **Claude.ai Subscriptions** (Free / Pro / Max)
2. **Claude API** (Pay-as-you-go)

---

# 🧾 1. Claude.ai Subscriptions

## Free Plan
- 💵 Cost: **$0/month**
- Access to Sonnet and Haiku models
- Web, iOS, Android, and desktop
- Text, image, file analysis, web search
- ⚠️ Daily usage limits apply (not published officially)

## Pro Plan
- 💵 Cost: **20 USD/month** (or **~17 USD/month** billed annually at $200 upfront)
- 👤 Designed for: Individual power users

### ✅ What You Get
- ~5x more usage than Free tier
- Access to Opus, Sonnet, and Haiku models
- Claude Code (terminal), file creation, code execution
- Unlimited projects
- Cross-conversation memory
- Google Workspace integration
- Research access
- Priority access during peak times

### ⚠️ Key Limitations
- ❌ No API access — cannot be used in your own applications
- ❌ Usage limits still apply (not unlimited)

---

## Max Plan
- 💵 Cost: **100 USD/month** (5x Pro usage) or **200 USD/month** (20x Pro usage)
- 👤 Designed for: Heavy users, researchers, developers who exceed Pro limits daily

### ✅ What You Get
- Everything in Pro
- Priority access to new features and models
- Claude in PowerPoint
- Higher output limits

### ⚠️ Key Limitation
- ❌ Still not API access — chat interface only

---

# ⚙️ 2. Claude API (Usage-Based Pricing)

- 💰 No monthly subscription fee
- 📊 Pay per **token** (input + output separately)
- Billed per **million tokens (MTok)**

## 💵 Current Pricing (per 1M tokens) — as of April 2026

| Model | Input (USD) | Output (USD) |
|-------|-------|--------|
| Claude Haiku 4.5 | 1.00 | 5.00 |
| Claude Sonnet 4.6 | 3.00 | 15.00 |
| Claude Opus 4.6 | 5.00 | 25.00 |

### 💡 Cost-Saving Options
| Option | Saving |
|--------|--------|
| Batch API (async, non-urgent tasks) | 50% off all models |
| Prompt Caching (repeated system prompts) | Up to 90% off cached input tokens |

---

## 🧠 How Billing Works

You are charged separately for:
- **Input tokens** — your prompt + context
- **Output tokens** — Claude's response

### 👉 Example Calculation (Opus 4.6)
- 50K input tokens → 50K ÷ 1,000,000 × 5.00 USD = **0.25 USD**
- 15K output tokens → 15K ÷ 1,000,000 × 25.00 USD = **0.375 USD**
- **Total ≈ 0.625 USD**

---

## ✅ What You Get
- Full developer control — build apps, chatbots, agents, RAG pipelines
- Access via Anthropic API, AWS Bedrock, Google Vertex AI, Azure
- Fine-grained cost control

## ⚠️ Key Limitations
- ❌ No fixed monthly cap → costs can grow quickly without monitoring
- ❌ **Rate limits apply** (requests/min and tokens/min per tier — not truly infinite)
- ❌ Requires coding / integration knowledge

---

# 🧠 Simple Analogy
- **Claude Pro/Max** = Hiring a personal assistant (fixed monthly retainer)
- **Claude API** = Renting an AI engine by the hour for your product

---

# 🎯 When to Use What?

## Use Claude Pro if:
- You are chatting, researching, writing, or coding personally
- You want predictable monthly cost
- You do not need to build applications

## Use Claude Max if:
- You hit Pro limits regularly
- You need maximum usage without switching to API

## Use Claude API if:
- You are building apps (chatbot, RAG, agents, automation)
- You need backend integration or programmatic access
- You want full control over prompts, models, and logic
- You can manage variable costs with monitoring

---

# 🚀 Key Takeaway
- **Pro/Max = Fixed cost, usage-capped, human-facing chat**
- **API = Variable cost, rate-limited (scales with tier), developer-focused**

✅ Choose based on **your use case, not just price**

# 🚀 Step 3 — First Claude API Call

````python
from google.colab import userdata

import anthropic

client = anthropic.Anthropic(api_key=userdata.get('MY_API_KEY'))

message = client.messages.create(
    model="claude-haiku-4-5-20251001",
    max_tokens=50,
    messages=[
        {
            "role": "user",
            "content": "Who is Arul Benjamin Chandru?",
        }
    ],
)
print(message)
````


---

# 🧩 Understanding the Code

### 🔹 model
Defines which Claude model to use.

Refer : https://platform.claude.com/docs/en/about-claude/models/overview

### 🔹 messages
Conversation format:
- `user` → your input
- `assistant` → model response

### 🔹 max_tokens
Limits response length.

### 🔹 Rate Limits
Limits help to mitigate against misuse and manage API capacity. Rate limits restrict API usage frequency over a certain period of time.

Refer : https://platform.claude.com/settings/limits

# 🎛️ Controlling Output

## 🔥 Temperature

```
temperature=0.7
```

Controls randomness:

| Value | Behavior |
|------|--------|
| 0.0 | Deterministic |
| 0.3 | Stable |
| 0.7 | Balanced |
| 1.0 | Creative |

---

## 🎯 Top-p (Nucleus Sampling)

```
top_p=0.9
```

Controls probability filtering.

- Temperature → randomness  
- Top-p → word selection range  



---

## 🧪 Example with Parameters



````python
response = client.messages.create(
    model="claude-haiku-4-5-20251001",
    max_tokens=150,
    temperature=0.0,
    messages=[
        {"role": "user", "content": "Write a short story about AI"}
    ]
)

print(response.content[0].text)
````

````python
response = client.messages.create(
    model="claude-haiku-4-5-20251001",
    max_tokens=150,
    top_p=0.9,
    messages=[
        {"role": "user", "content": "Write a short story about AI"}
    ]
)

print(response.content[0].text)
````

````python
import anthropic

def call_claude(prompt):
    client = anthropic.Anthropic(
       api_key=userdata.get('FREE_API')
    )

    response = client.messages.create(
        model="claude-haiku-4-5-20251001",
        max_tokens=200,
        top_p=0.9,
        messages=[
          {"role": "user", "content": prompt}
        ]
    )

    text_output = response.content[0].text
    usage = response.usage

    print("\n--- Claude Response ---")
    print(text_output)

    print("\n--- Token Usage ---")
    print(f"Input Tokens: {usage.input_tokens}")
    print(f"Output Tokens: {usage.output_tokens}")

    return text_output


user_prompt = input("Enter your prompt: ")
call_claude(user_prompt)
````
