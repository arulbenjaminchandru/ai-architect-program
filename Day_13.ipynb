## Section 1: What Is CLAUDE.md and Why It Exists


### The Core Problem CLAUDE.md Solves

Every time you start a new Claude Code session, Claude Code begins without memory of previous
sessions. It does not remember what you worked on yesterday, what conventions your team uses,
or what constraints matter to your project.

**The Session Memory Problem:**

```
Session 1 (Monday):
  You tell Claude Code: "We use camelCase for variables"
  Claude Code: Uses camelCase correctly
  Session ends.

Session 2 (Tuesday):
  New session — no memory of Monday
  Claude Code: Uses snake_case (its default)
  You: Have to tell it again
  Repeated every single session

This is the problem CLAUDE.md solves.
```

**The Solution:**

```
CLAUDE.md is a file in your project directory.
Claude Code reads it at the start of every session.
It is permanent, written project context.

Instead of:
  You → tell Claude Code conventions every session

With CLAUDE.md:
  Claude Code reads file → knows conventions
  Claude Code reads file → knows constraints
  Claude Code reads file → knows patterns
  Every session, automatically
```


### What CLAUDE.md Is (Precise Definition)

```
CLAUDE.md is:
  - A Markdown file (.md format)
  - Located in your project root (and/or home directory)
  - Read by Claude Code at session start
  - A way to provide persistent context
  - Written by you (or your team)
  - Version-controlled with your project (typically)

CLAUDE.md is NOT:
  - A configuration file that executes code
  - A replacement for code documentation
  - A file that other tools read (only Claude Code reads it)
  - An automatic file (you write it, Claude doesn't generate it)
  - A guarantee of behavior (it influences Claude, not programs it)
```

**Analogy — The Briefing Document:**

```
Imagine hiring a contractor:

  Without briefing:
    - Asks questions every day
    - Makes assumptions that may be wrong
    - Delivers work that needs rework

  With a briefing document:
    - Reads it before starting
    - Understands your standards
    - Follows your patterns
    - Delivers aligned work

CLAUDE.md is your briefing document for Claude Code.
```

### Where CLAUDE.md Files Live

Claude Code checks multiple locations for CLAUDE.md files:

```
1. Home directory (~/ or ~/.claude/)
   Purpose: Personal preferences that apply to all projects
   Contains: Your personal coding style, tools you prefer,
             general preferences across all your work

2. Project root (/your-project/CLAUDE.md)
   Purpose: Project-wide context for everyone on the team
   Contains: Architecture, conventions, constraints,
             commands, skills registry

3. Subdirectory CLAUDE.md (/your-project/src/CLAUDE.md)
   Purpose: Module-specific context
   Contains: Context specific to one part of the codebase
             (only loaded when working in that directory)
```

**Why Multiple Locations Matter:**

```
Personal CLAUDE.md (home directory):
  You prefer: "Always explain your changes before making them"
  Applies to: Every project you work on

Project CLAUDE.md (project root):
  Team agreed: "We use PostgreSQL, never suggest MongoDB"
  Applies to: Everyone on the team using Claude Code

Subdirectory CLAUDE.md:
  Payment module: "Extra caution here — all changes need tests"
  Only applies: When working inside /src/payments/

These compose: All relevant CLAUDE.md files are read together.
```

---

### Key Takeaways — Section 1

- CLAUDE.md is a **Markdown file** read by Claude Code at the start of every session
- It solves the **session memory problem** — Claude Code cannot remember previous sessions on its own
- CLAUDE.md **influences** Claude Code; it does not program or hard-lock its behaviour
- There can be **multiple CLAUDE.md files** — home directory, project root, and subdirectories
- It should be **version-controlled** with your project so the whole team benefits

## Section 2: How Claude Code Uses CLAUDE.md


### The Reading Process (Verified Behavior)

**What happens when you start Claude Code:**

```
Step 1: You run "claude" in your project directory

Step 2: Claude Code checks for CLAUDE.md files
  - Checks home directory (~/)
  - Checks project root
  - Checks subdirectories relevant to current context

Step 3: Claude Code reads CLAUDE.md content
  - Content becomes part of Claude's context
  - Same as if you had typed this content in the conversation
  - But it happens automatically at session start

Step 4: Claude Code responds to your requests
  - With awareness of CLAUDE.md content
  - Uses conventions you specified
  - Respects constraints you defined
  - Uses commands you defined
  - Knows about skills you registered

Important accuracy note:
  CLAUDE.md content goes into Claude's context window.
  It is treated like a system prompt or initial context.
  Claude Code reads it — it does not "learn" it permanently.
  Each new session, CLAUDE.md is re-read from scratch.
```


### What "Context Window" Means Here

```
Claude (the AI model) processes text in a context window.
The context window is the total amount of text Claude
can "see" at one time during a conversation.

CLAUDE.md content occupies part of this context window.
Your conversation also occupies the context window.

Implication:
  Very long CLAUDE.md files use up context window space.
  Shorter, focused CLAUDE.md files leave more room
  for your actual conversation and code.

This is why Skills (Skill.md) exist:
  They allow you to split context across files
  and load only what is needed for the current task.
```


### From Instructions to Behavior

**Without CLAUDE.md:**

```
You ask: "Add error handling to this function"

Claude Code:
  - Does not know your error handling pattern
  - Uses a generic pattern (try/catch with console.error)
  - Or asks: "How do you want to handle errors here?"
  - Suggestion may not match your codebase
  - You have to modify it or ask again
```

**With CLAUDE.md (containing error handling pattern):**

```
CLAUDE.md states:
  "Error handling: All async functions use try/catch.
   Errors are thrown as CustomError class instances.
   Include error code and user-facing message.
   Log with our logger utility: logger.error()"

You ask: "Add error handling to this function"

Claude Code:
  - Knows your pattern from CLAUDE.md
  - Applies CustomError, includes error code, uses logger
  - Matches your existing codebase
  - No clarifying questions asked, no rework needed
```

> **Important distinction:** CLAUDE.md does not *program* Claude Code.
> CLAUDE.md *informs* Claude Code. Claude then applies that information when generating output.

---

### Key Takeaways — Section 2

- CLAUDE.md content is loaded into Claude Code's **context window** at session start
- It behaves like a system prompt — Claude Code reads and applies it, not stores it permanently
- Long CLAUDE.md files **consume context window space**, leaving less room for actual work
- Without CLAUDE.md, Claude Code defaults to generic patterns; with it, output aligns to your project
- Skills (Skill.md) exist to keep the main CLAUDE.md lean by splitting context into domain-specific files

## Section 3: Project Context — The Five Categories


### Why Context Categories Matter

Not all information Claude Code needs is the same kind of information. Organising CLAUDE.md
by category makes it easier to write, easier to read, and easier to maintain.
The five categories below cover everything Claude Code typically needs to understand a project.


### Category 1: Architectural Context

**Definition:** Describes how your project is structured at the highest level — what the major
components are, how they relate to each other, and where code lives.

**What to include:**

```
1. Project structure
   - Is it a monolith, microservices, or something else?
   - Major directories and what they contain
   - Where to create new files for different things

2. Component relationships
   - Which components depend on which
   - What talks to what
   - Data flow direction

3. Technology boundaries
   - Where is Node.js used vs Python vs other languages?
   - What parts use React vs server-rendered HTML?
   - What is the database layer?

4. Key files and directories
   - Where is the entry point?
   - Where does business logic live?
   - Where do tests live?
```

**Why Claude Code needs this:**

```
Without architecture context:
  "Create a user profile feature"
  → Claude Code creates files in the wrong location
  → You spend 20-30 minutes moving and reorganising

With architecture context:
  CLAUDE.md says: "New features go in src/features/<name>/"
  → Claude Code creates src/features/user-profile/ correctly
  → Correct first time, zero rework
```


### Category 2: Domain Context

**Definition:** The business knowledge Claude Code needs — what problem you are solving,
for whom, and under what rules.

**What to include:**

```
1. Business domain
   - E-commerce? Healthcare? Finance? SaaS?
   - Core entities: Users, Orders, Products?
   - Key workflows

2. Domain-specific rules
   - Business logic that must be respected
   - Calculations that must be precise
   - Rules that are non-negotiable

3. Domain terminology
   - What does "account" mean in YOUR context?
     (User account or financial account?)
   - What does "order" mean?
     (Purchase order or sort order?)

4. Regulatory context (if relevant)
   - Healthcare → HIPAA
   - Finance → PCI, SOX
   - EU data → GDPR
```

**Why Claude Code needs this:**

```
Without domain context:
  "Optimise the order total calculation"
  → Claude Code simplifies the formula
  → Breaks tax rules, discount rules, currency rounding
  → Potential legal issue

With domain context:
  CLAUDE.md says: "Order calculations are financial.
                   Rounding must follow [specific rule].
                   Never simplify this code."
  → Claude Code understands the constraint
  → Suggests only safe optimisations
```


### Category 3: Technical Context

**Definition:** The concrete detail of what technologies, versions, and patterns you use.

**What to include:**

```
1. Technology stack
   - Language and version: Python 3.11, Node.js 20
   - Framework and version: React 18, Express 4, Django 5
   - Database and version: PostgreSQL 16, MongoDB 7
   - Key libraries and their versions

2. Configuration approach
   - How is configuration managed? (env vars, config files)
   - Where are environment variables defined?
   - Pattern for secrets management

3. Testing approach
   - Test framework: Jest, pytest, RSpec
   - Where do tests live?
   - Testing patterns

4. Build and run approach
   - How to start the dev server
   - How to run tests
   - How to build for production
```

### Category 4: Team and Process Context

**Definition:** How your team works — the human processes around the code.

**What to include:**

```
1. Team information
   - Team size (relevant to complexity of solutions)
   - Who reviews what (approval processes)

2. Development process
   - Branch strategy: gitflow? trunk-based?
   - PR/review requirements
   - Deployment approach

3. Code review standards
   - What triggers a rejection?
   - What is required to merge?
   - Automated checks that must pass
```

### Category 5: Historical and Decision Context

**Definition:** Why things are the way they are — decisions made, paths tried, lessons learned.

**What to include:**

```
1. Key decisions and rationale
   - Why PostgreSQL instead of MongoDB?
   - Why REST instead of GraphQL?

2. Tried and rejected approaches
   - What was tried and why it was abandoned
   - Why you should not suggest those approaches again

3. Known technical debt
   - What areas are messy and why
   - What should not be refactored right now

4. Gotchas and surprises
   - Things that look like they should work differently
   - Non-obvious dependencies
```

---

### Key Takeaways — Section 3

- There are **five context categories**: Architectural, Domain, Technical, Team/Process, Historical
- **Architectural context** prevents Claude Code from putting files in the wrong place
- **Domain context** protects business logic from well-intentioned but incorrect "optimisations"
- **Technical context** ensures Claude Code uses the right libraries, patterns, and tools
- **Historical context** prevents Claude Code from re-suggesting approaches the team already rejected

## Section 4: Conventions — Teaching Claude Your Patterns


### What Conventions Are in This Context

Conventions are the agreed-upon ways of doing things in your project. They are not rules Claude
Code enforces at a system level — they are patterns Claude Code learns from CLAUDE.md and applies
when generating code.


### Convention Type 1: Naming Conventions

**What naming conventions cover:**

```
Variables:
  camelCase:  let userName, let orderTotal
  snake_case: user_name, order_total
  Choice is team preference — document it explicitly.

Functions and methods:
  Verb-first: calculateTotal(), processPayment(), getUserById()
  Prefix patterns:
    handle  → event handlers
    get/set → accessors
    is/has/can → booleans
    on      → callbacks

Classes:
  PascalCase is standard: UserService, OrderProcessor
  Some teams add suffixes: UserRepository, OrderController

Files:
  kebab-case:  user-service.ts, order-processor.ts
  PascalCase:  UserService.ts, OrderProcessor.tsx (React)
  snake_case:  user_service.py (Python convention)

Constants:
  SCREAMING_SNAKE_CASE: MAX_RETRY_COUNT, DEFAULT_TIMEOUT

Git branches:
  feature/description, bugfix/description, hotfix/description
```

**Why Claude Code needs explicit naming conventions:**

```
Claude Code knows general conventions per language.
What it does NOT know without CLAUDE.md:
  - Your specific prefix choices
  - Whether you use camelCase or snake_case in JS
  - Your naming pattern for React components
  - Your branch naming convention

Without CLAUDE.md: Claude Code defaults to common patterns
                   which may or may not match YOUR project

With CLAUDE.md:    Claude Code follows your patterns
                   New code matches existing code
                   No rename needed later
```


### Convention Type 2: Code Structure Conventions

**What structure conventions cover:**

```
Class organisation example:
  1. Static properties
  2. Instance properties
  3. Constructor
  4. Public methods (alphabetical or logical order)
  5. Private methods
  6. Static methods

Import ordering example:
  1. Standard library imports
  2. Third-party package imports
  3. Internal module imports
  4. Relative imports
  (Blank line between each group)

Error handling style options:
  Option A: Throw exceptions, catch at boundary
  Option B: Return error objects { data, error }
  Option C: Use Result types (functional style)
  Each is valid — document yours.

Documentation style options:
  JSDoc format:  /** @param {string} name */
  Plain comment: // Takes a user name string
  Docstring:     """Takes a user name string"""
```

### Convention Type 3: Architectural Conventions

**What architectural conventions cover:**

```
Feature organisation:
  By layer:    src/controllers/, src/services/, src/models/
  By feature:  src/users/, src/orders/, src/products/
               (each containing controller, service, model)
  Both valid — document which you use.

Where business logic lives:
  Controller-heavy: Logic in controllers (simpler apps)
  Service layer:    Controllers route, services contain logic
  Domain layer:     Rich domain objects with behaviour

API response format:
  Document your standard shape, e.g.:
  { data: {}, errors: [], meta: {} }

Testing structure:
  Co-located: tests next to source files
  Central:    tests/ directory mirroring src/
```


### Three Principles of Effective Convention Writing

**Principle 1: Be explicit, not vague**

```
Vague (unhelpful):
  "Follow clean code principles"
  "Use good naming"
  "Write readable code"

Explicit (helpful):
  "Variable names: camelCase in TypeScript
   Exception: constants use SCREAMING_SNAKE_CASE
   Exception: private class properties use _camelCase prefix"

Vague gives Claude Code nothing to act on.
Explicit gives Claude Code a clear pattern to follow.
```

**Principle 2: Show the pattern AND an example**

```
Pattern alone:
  "Functions that return booleans use is/has/can prefix"

Pattern with example:
  "Functions that return booleans use is/has/can prefix.
   Examples:   isUserActive(), hasPermission(), canAccessResource()
   NOT:        checkUser(), userStatus(), active()"

Examples prevent ambiguity. Claude Code can pattern-match
on examples even when the rule is slightly ambiguous.
```

**Principle 3: Document exceptions**

```
Rule without exception:
  "All functions must have type annotations"
  → Claude Code applies this everywhere, even where it does not fit.

Rule with exception:
  "All functions must have type annotations.
   Exception: small utility scripts (under 30 lines)
   Exception: test files (types are often noisy there)
   Exception: dynamically typed data from external APIs"

Exceptions make rules realistic and useful.
Without exceptions, Claude Code may over-apply rules.
```

---

### Key Takeaways — Section 4

- Conventions in CLAUDE.md are **pattern instructions**, not hard technical rules
- Always be **explicit** — vague principles like "clean code" give Claude Code nothing to act on
- Always provide **examples** of correct usage (and incorrect where helpful)
- Always document **exceptions** — rules without exceptions get over-applied
- Conventions span naming, code structure, error handling, imports, and architecture placement

## Section 5: Constraints — What Claude Code Must Not Do


### What Constraints Are (Precise Definition)

Constraints in CLAUDE.md are instructions telling Claude Code what to avoid, what must not change,
and what requires human approval before proceeding.

> **Important accuracy note:** Constraints in CLAUDE.md do not technically prevent Claude Code
> from doing something at a system level. They are instructions that Claude Code reads and respects.
> Claude Code is designed to follow these instructions — but this is Claude following instructions,
> not a hard technical lock. Always review Claude Code's changes, especially in critical areas.
> Never rely solely on CLAUDE.md constraints for security.


### Constraint Categories

**Hard Constraints — Must Not Violate:**

```
These are constraints where the consequences of violation
are severe: security risk, data loss, legal issue, or
broken functionality that affects users.

Examples:
  "Never store passwords in plaintext.
   Reason: Security requirement. Violating this is a breach."

  "Never modify the payments/ directory without review.
   Reason: PCI compliance. Changes must be peer-reviewed."

  "Never remove audit logging from user actions.
   Reason: Legal requirement for compliance."

  "Never delete records from the users table.
   Reason: GDPR soft-delete requirement. Use is_deleted flag."

How to write them:
  - State the constraint clearly
  - State the REASON (Claude Code responds to reasoning)
  - State who to consult if change is needed
  - Use strong language: never, must not, do not
```

**Soft Constraints — Should Follow, Can Break with Explanation:**

```
These represent best practices or team preferences
but have legitimate exceptions.

Examples:
  "Functions should be under 50 lines.
   Reason: Readability.
   Exception: State machines and sequential workflows
   may be longer. Comment why if you exceed this."

  "Aim for 80% test coverage.
   Reason: Quality and confidence.
   Exception: Generated code, simple getters, trivial mappers."

How to write them:
  - State the guideline and reason
  - State known exceptions
  - Use moderate language: should, aim for, prefer
```

**Contextual Constraints — Apply in Specific Situations:**

```
Examples:
  "Do not refactor the legacy/ directory.
   Reason: Code there is stable but not well-understood.
   Exception: If explicitly asked to, add thorough tests first."

  "Changes to shared API contracts (api/types.ts)
   affect both frontend and backend.
   Always note potential breaking changes."
```


### Writing Constraints for Maximum Effectiveness

**Key finding:** Claude Code responds better to constraints when the reason is given,
the constraint is specific, exceptions are acknowledged, and consequences are clear.

**Ineffective constraint:**

```
"Be careful with the database"
→ Too vague. Careful how? This gives Claude Code nothing actionable.
```

**Effective constraint:**

```
"Never drop database columns in migrations.
 Reason: We have multiple servers that deploy at different times.
 Dropping a column breaks servers not yet deployed.
 Instead: First migration makes the column optional (nullable).
          Second migration (next deploy cycle) removes the column."
```

The second version is specific, gives the reason, and shows the correct alternative.

---

### Key Takeaways — Section 5

- Constraints are **instructions**, not hard system locks — always review Claude Code's changes
- Use **hard constraints** (never/must not) for security, compliance, and data integrity issues
- Use **soft constraints** (should/prefer/aim for) for best practices with legitimate exceptions
- Always include the **reason** — constraints without reasons feel arbitrary and get weighted poorly
- Use **contextual constraints** for directories or code areas that need extra care
- Version control is your **true safety net**; CLAUDE.md constraints are a useful layer, not the only one

## Section 6: Custom Slash Commands — Consistency and Shortcuts


### What Custom Slash Commands Are

**Precise definition:**

```
Custom slash commands in CLAUDE.md are shortcuts you define
that expand into longer prompts or instructions when you
type them in a Claude Code conversation.

They are defined in CLAUDE.md and used by you (the developer)
when talking to Claude Code.

They are NOT:
  - Shell scripts that execute automatically
  - System commands like /bin/bash commands
  - VS Code commands
  - Macros that run without Claude Code involvement

They ARE:
  - Prompt shortcuts
  - Consistency tools (same instruction every time)
  - Workflow helpers (complex instructions in a short command)
```

**How they actually work:**

```
You define in CLAUDE.md:

  /run-tests: Run the test suite using "npm test" and report results.
              Focus on any failing tests and suggest fixes.

When you type "/run-tests" in Claude Code:
  Claude Code reads this as the full instruction.
  Claude Code then runs npm test (via terminal access).
  Claude Code then reports results as specified.

The slash command is a shortcut to a full instruction.
The action (running npm test) happens because Claude Code
is an agentic tool that can execute shell commands.
The slash command itself is not a shell script.
```


### Why Slash Commands Are Valuable

**Consistency:**

```
Without command:
  You explain what you want each time.
  The explanation varies slightly each session.
  Claude Code gets slightly different instructions.
  Results vary.

With command:
  /code-review always means exactly the same thing.
  Every code review follows the same checklist.
  Consistent quality every time.
```

**Speed:**

```
Without command:
  "Can you review this for security issues,
   performance problems, make sure it follows
   our CLAUDE.md conventions including naming patterns
   and error handling, and check for missing tests?"

With command:
  /code-review
  (Same instruction, one word.)
```

**Shareability:**

```
Team agrees on what /deploy-check should do.
Everyone uses the same command.
Same process, regardless of who runs it.
```


### Types of Slash Commands to Define

**Development workflow commands:**

```
/run-tests
  Instruction: Run the test suite. Report pass/fail counts.
  For any failures, show the test name and error.
  Suggest potential causes.

/check
  Instruction: Run tests, linting, and type checking.
  Report pass/fail for each, then detail failures.

/lint
  Instruction: Run the project linter as described in CLAUDE.md.
  Report issues by file and line number. Suggest fixes.
```

**Code quality commands:**

```
/code-review
  Instruction: Review this code for:
  1. Security concerns
  2. Performance issues
  3. Violations of conventions in this CLAUDE.md
  4. Missing or weak error handling
  5. Missing tests for new logic
  Format: One section per category. Only report issues found.

/security-check
  Instruction: Review this code specifically for security concerns.
  Check against OWASP Top 10 issues. Be thorough and conservative.
```

**Workflow consistency commands:**

```
/new-feature <name>
  Instruction: I want to create a feature called <name>.
  Before doing anything:
  1. List all files you plan to create and their purpose
  2. List any existing files you will modify
  3. Estimate scope: small / medium / large
  4. Ask me to confirm before proceeding

/refactor <target>
  Instruction: I want to refactor <target>.
  Before making changes:
  1. Summarise current approach
  2. Explain proposed changes and rationale
  3. Note any risks or concerns
  4. Ask me to confirm before proceeding
```


### Three Principles of Effective Slash Command Design

**Principle 1: Commands should be memorable and intuitive**

```
Good:    /run-tests, /code-review, /new-feature, /security-check
Poor:    /rt, /do-the-thing, /helper, /x

Rule: If a new team member sees the command name,
they should understand roughly what it does.
```

**Principle 2: Specify the output format**

```
Good definition:
  /run-tests: Run "npm test". Output format:
  - First line: PASS or FAIL summary
  - List any failing test names with their errors
  - End with: number passed, failed, total

Poor definition:
  /run-tests: Run the tests

Why it matters:
  Consistent output format means you can
  quickly scan results every time.
```

**Principle 3: Commands that change things should confirm first**

```
Example:
  /new-feature <name>:
    First: Describe what files will be created
    Then:  Ask "Does this plan look right?"
    Only then: Create files

Why:
  Claude Code can make many changes quickly.
  A confirmation step prevents unwanted large changes.
  Good practice for any agentic tool.
```

---

### Key Takeaways — Section 6

- Slash commands are **prompt shortcuts** defined in CLAUDE.md, not shell scripts or macros
- They expand to the full instruction you wrote when typed in a Claude Code conversation
- Primary benefits: **consistency**, **speed**, and **team-wide shareability**
- Always **specify output format** in the command definition for predictable results
- Any command that modifies files should **ask for confirmation** before proceeding
- Keep command names **intuitive** — new team members should understand them at a glance

## Section 7: Claude Skills and Skill.md — Modular Knowledge


### What Skills Are (Verified Definition)

**The problem Skills solve:**

```
A large project CLAUDE.md might need to cover:
  - Testing patterns and framework
  - Deployment process and commands
  - Database patterns and migration approach
  - API design and versioning
  - Frontend component patterns
  - Security patterns and requirements
  - Documentation standards

If all of this is in one CLAUDE.md:
  - The file becomes very long
  - Claude Code loads all of it every session
  - Much of it is irrelevant to the current task
  - Context window fills up quickly
  - Less room for your actual code and conversation

Skills (Skill.md files) solve this by splitting context
across multiple files, each covering one domain.
```

**What Skill.md files are:**

```
Skill.md files are specialised CLAUDE.md-like files
that cover a specific domain or capability.

They are:
  - Referenced from your main CLAUDE.md
  - Read by Claude Code when relevant to the task
  - Focused on one domain (testing, deployment, etc.)
  - More detailed than what belongs in CLAUDE.md

The pattern:
  CLAUDE.md: Overview, conventions, constraints, slash commands,
             AND a Skills section pointing to Skill.md files

  Skill.md files: Deep domain-specific guidance
```


### How Claude Code Uses Skill.md Files

```
The pattern depends on how you write CLAUDE.md.

Option A: Always-load skills
  CLAUDE.md lists skills that should always be loaded.
  Claude Code reads them every session.
  Good for: Core skills all developers use daily.

Option B: Task-triggered loading
  CLAUDE.md instructs: "When working on tests, read testing/SKILL.md"
  Claude Code reads the skill when starting test work.
  Good for: Domain skills used less frequently.

Option C: Explicit loading
  You type: "Please read testing/SKILL.md before we start"
  Claude Code reads it on your request.
  Good for: Skills you use infrequently.

Option D: Slash command triggered
  /write-tests: "Read testing/SKILL.md then write tests for..."
  Good for: Predictable repeated workflows.
```

> **Accuracy note:** The "automatic detection" of which skill to load is NOT an automatic
> system feature. It happens because you wrote CLAUDE.md to instruct it, or because you
> trigger it explicitly. Design your CLAUDE.md to make skill loading explicit and predictable,
> not to rely on Claude inferring it.


### What Skill Categories to Create

```
Technical domain skills:
  testing/SKILL.md      Testing approach, frameworks, patterns
  database/SKILL.md     Database patterns, migration approach
  api/SKILL.md          API design, endpoints, versioning
  frontend/SKILL.md     UI patterns, components, state management
  deployment/SKILL.md   Deployment steps, environments, process
  security/SKILL.md     Security requirements, patterns, checks

Domain/business skills:
  billing/SKILL.md          How billing logic works in your system
  notifications/SKILL.md    Notification patterns
  auth/SKILL.md             Authentication and authorisation

Process skills:
  code-review/SKILL.md    What the review process looks like
  debugging/SKILL.md      How to debug common issues in this project
  onboarding/SKILL.md     How to get started (for Claude Code and humans)

When to create a new Skill.md:
  - A section of CLAUDE.md is getting very long
  - A domain has many specific patterns to document
  - The domain is only relevant to some developers
  - You want to update domain knowledge without touching main CLAUDE.md
```


### What Goes in a Skill.md File

**The key difference from CLAUDE.md:**

```
CLAUDE.md is broad:       Covers the whole project, brief per section
Skill.md is deep:         Covers one domain in detail, can be thorough

This means:
  In CLAUDE.md:           "We use Jest for testing"
  In testing/SKILL.md:    Everything about testing in your project
                          (test structure, naming, mocking patterns,
                           database setup, coverage requirements,
                           CI integration, common gotchas)
```

**Skill.md structure:**

```
# [Skill Name] for [Project Name]

## When to Read This
  Read when: [specific triggers/situations]

## Overview
  [2-3 sentence description of this domain in your project]

## Core Concepts
  [Key concepts — what Claude Code needs to know about
   how YOUR project handles this domain]

## Patterns
  [Established patterns used in this domain.
   Describe the pattern, not just name it.
   Include rationale where helpful.]

## Conventions
  [Naming and structural conventions specific to this domain]

## Constraints
  [What NOT to do in this domain, with reasons]

## Tools and Commands
  [What tools are used, how to invoke via slash commands]

## Common Gotchas
  [Non-obvious things that have surprised people]

## Examples
  [References to good examples in the codebase.
   "See src/features/auth/ for a good example"
   — not pasted code that goes out of date]
```

> **Tip:** Use file references ("see src/features/auth/AuthService.ts") rather than pasting code.
> Code examples go stale; file references stay valid as implementations evolve.

### Key Takeaways — Section 7

- Skill.md files solve the **context bloat problem** by splitting domain knowledge into focused files
- They are **referenced from CLAUDE.md** and loaded when needed — not auto-detected automatically
- Loading happens via always-load, task-triggered instruction, explicit request, or slash command
- A Skill.md file is **deeper but narrower** than CLAUDE.md — go into full detail within one domain
- Use **file references** rather than pasting code examples that will go stale
- Create a new Skill.md when any CLAUDE.md section starts getting too long to be useful inline

## Section 8: CLAUDE.md Structure — The Full Template


### The Eight Core Sections

A well-structured CLAUDE.md for a typical project:

```
1. Project Overview       (brief — 5-10 lines)
2. Architecture           (20-40 lines)
3. Tech Stack             (5-15 lines)
4. Conventions            (40-80 lines)
5. Constraints            (30-60 lines)
6. Custom Commands        (30-60 lines)
7. Skills Registry        (15-30 lines)
8. Development Workflow   (20-40 lines)

Rough total: under 200 lines for most projects
```


### Section-by-Section Guide

**Section 1: Project Overview**

```
What to include:
  - 1-3 sentences: what the project does
  - Who the users are
  - The most important thing Claude Code should know
  - Critical context that shapes all other decisions

What to exclude:
  - Installation instructions (that is README.md)
  - Project history (unless it affects decisions)
  - Roadmap (unless constraints come from it)

Target: 5-10 lines
```

**Section 2: Architecture**

```
What to include:
  - Project structure (text or ASCII diagram)
  - Major directories and what they contain
  - Where to create files for different things
  - Key architectural decisions that affect where code goes

ASCII directory tree works well:

  project-root/
  ├── src/                 Application code
  │   ├── features/        Feature modules (by domain)
  │   ├── shared/          Code shared across features
  │   └── infrastructure/  DB, external APIs, file system
  ├── tests/               Test files (mirror src/ structure)
  ├── scripts/             Build, migration, utility scripts
  └── docs/
      └── skills/          Skill.md files

Target: 20-40 lines
```

**Section 3: Tech Stack**

```
What to include:
  - Primary language and version
  - Main framework and version
  - Database and client library
  - Key libraries (ones that matter to how code is written)
  - Test framework
  - Build tools

What NOT to include:
  - Every package.json dependency
  - Dev tools Claude Code does not need to know about

Format example:
  Backend:  Node.js 20, Express 4.18
  Database: PostgreSQL 16, via pg library (no ORM)
  Frontend: React 18, TypeScript 5
  Testing:  Jest 29, Supertest

Target: 5-15 lines
```

**Section 4: Conventions**

```
What to include:
  - Naming conventions with examples
  - Code structure conventions
  - Error handling approach
  - Documentation approach
  - Import organisation

Target: 40-80 lines
```

**Section 5: Constraints**

```
Structure suggestion:
  ## Constraints

  ### Never Do These
  (Hard constraints with reasons and who to consult)

  ### Prefer / Avoid
  (Soft constraints with context and exceptions)

  ### Sensitive Areas
  (Code that needs extra care and why)

Target: 30-60 lines
```

**Section 6: Custom Commands**

```
Format:
  ## Custom Commands

  /run-tests
    [full instruction with output format]

  /code-review
    [full instruction with checklist]

Target: 30-60 lines (depends on number of commands)
```

**Section 7: Skills Registry**

```
Format:
  ## Skills

  Available skills in docs/skills/:

  testing/SKILL.md
    Covers: Jest setup, testing patterns, mocking approach
    Load when: Writing or reviewing tests

  deployment/SKILL.md
    Covers: Deployment process, environments, rollback
    Load when: Working on deployment or release tasks

  database/SKILL.md
    Covers: Database patterns, migration conventions
    Load when: Writing migrations or database queries

Target: 15-30 lines
```

**Section 8: Development Workflow**

```
What to include:
  - How to start dev environment
  - How to run tests
  - How to build
  - Branching convention
  - Commit message convention
  - PR process

Why Claude Code needs this:
  Claude Code may start your dev server.
  Claude Code needs to run tests correctly.
  Claude Code writes commit messages following your convention.

Target: 20-40 lines
```

---

### Key Takeaways — Section 8

- A well-structured CLAUDE.md has **eight sections** covering overview, architecture, stack, conventions, constraints, commands, skills, and workflow
- Target **under 200 lines** for the core CLAUDE.md — move longer sections to Skill.md files
- Every section has a **clear target length** — stay within it to conserve context window space
- The **Skills Registry** section acts as a table of contents for your Skill.md files
- The **Development Workflow** section enables Claude Code to start servers, run tests, and commit using your project conventions

## Section 9: Skill.md Structure — Writing Effective Skills


### The Skill.md Template

```
# [Skill Name] for [Project Name]

## When to Read This
Read this skill when: [specific triggers/situations]
Do not load if:       [situations where this is irrelevant]

## Overview
[2-3 sentence description of this domain in your project.
 Not general knowledge — what is specific to YOUR project.]

## Core Concepts
[Key concepts Claude Code needs to understand.
 Only include what is specific to your project setup.]

## Patterns

### Pattern 1: [Name]
  [Description of the pattern]
  [When to use it]
  [Why this pattern was chosen over alternatives]

### Pattern 2: [Name]
  [Same structure]

## Conventions
[Naming and structural conventions specific to this domain]

## Constraints
[What NOT to do in this domain, with reasons]

## Tools and Commands
[What tools are used, how to invoke via slash commands]

## Common Gotchas
[Things that are non-obvious or have surprised people before]

## Examples
[References to good examples in the actual codebase.
 Use file paths, not pasted code:
 "See src/features/auth/AuthService.ts for token refresh pattern"]
```


### What Good Testing Skill.md Looks Like (Example)

```
# Testing Skill for [Project Name]

## When to Read This
Read this skill when: Writing tests, reviewing tests,
                      or debugging failing tests.

## Overview
This project uses Jest for unit and integration tests.
E2E tests use Playwright.

## Testing Approach
  - Unit tests:        Test individual functions in isolation
  - Integration tests: Test service layer with real test database
  - E2E tests:         Test user-facing workflows in browser
  - No snapshot tests (they break too easily with minor UI changes)

## Test File Conventions
  - Unit:         filename.test.ts, co-located with source file
  - Integration:  filename.integration.test.ts
  - E2E:          tests/e2e/feature-name.spec.ts

## Test Organisation Pattern
  describe('<ClassName or feature>', () => {
    describe('<methodName or scenario>', () => {
      it('should <expected behaviour>', ...)
      it('should <other expected behaviour>', ...)
    })
  })

## Database in Tests
  - Unit tests:         mock the database layer
  - Integration tests:  use test database (TEST_DATABASE_URL)
  - Always reset between tests using beforeEach hooks
  - Never use production database in tests

## Mocking Approach
  - External APIs:      always mock
  - Time:               use jest.useFakeTimers()
  - File system:        mock with memfs library
  - Internal services:  use real implementation
                        unless it has external dependencies

## Constraints
  - Never commit tests that hit external APIs without mocking
  - Test coverage must not drop below current baseline
  - E2E tests require running local stack (/start-stack)

## Common Gotchas
  - Using real API calls in tests (slow and flaky)
  - Not resetting state between tests (test pollution)
  - Testing implementation details instead of behaviour

## Examples
  See: src/features/auth/__tests__/AuthService.test.ts
  (Good example of service-level unit tests with mocking)
```

---

### Key Takeaways — Section 9

- Skill.md follows the same principles as CLAUDE.md but goes **deeper within one domain**
- Always include a **"When to Read This"** section so Claude Code and humans know when to load it
- Include **rationale** for patterns — not just what the pattern is, but why the team chose it
- Use **file references** to examples in the actual codebase rather than pasting code inline
- Cover **common gotchas** — these save time by preventing repeated mistakes
- Each Skill.md should be **self-contained** — someone reading it cold should get the full picture

## Section 10: Claude Code + VS Code — The Integrated Workflow


### How Claude Code and VS Code Work Together

**Clarifying the relationship:**

```
Claude Code (the CLI tool):
  - Runs in your terminal
  - Has access to your files, shell, and git
  - Reads CLAUDE.md files
  - Is the primary Claude Code interface

VS Code (the editor):
  - Shows your code
  - Has an integrated terminal
  - Can run Claude Code in its terminal
  - Has a Claude Code extension for tighter editor integration

The integration is:
  VS Code terminal → runs Claude Code CLI
  OR
  VS Code extension → surfaces Claude Code in the editor UI

Both approaches use the same underlying Claude Code tool.
The extension adds editor-level integration (inline suggestions, etc.).
```

**Two methods of working:**

```
Method 1: Terminal in VS Code
  1. Open VS Code
  2. Open project folder (File → Open Folder)
  3. Open integrated terminal (Terminal → New Terminal)
  4. Navigate to project root
  5. Run: claude
  6. Claude Code starts, reads CLAUDE.md
  7. Split view: editor on left, Claude Code terminal on right
  8. Claude Code reads/edits files visible in your editor
  9. Changes appear in editor in real-time

Method 2: VS Code Extension
  1. Install Claude Code VS Code extension
  2. Open project folder
  3. Access Claude Code via extension panel
  4. Extension provides tighter integration
     (inline suggestions, easier file context sharing)

Both methods:
  - Work with CLAUDE.md in exactly the same way
  - Have access to the same capabilities
  - Read the same CLAUDE.md and Skill.md files
```


### The Step-by-Step Workflow with CLAUDE.md

```
Step 1: Session starts
  You run "claude" in project terminal.
  Claude Code reads:
    - CLAUDE.md in project root
    - CLAUDE.md in home directory (if it exists)

  Claude Code now knows:
    - Project structure and architecture
    - Conventions to follow
    - Constraints to respect
    - Slash commands available
    - Skills that exist

Step 2: You give a task
  "I need to add a forgot-password flow"

  Claude Code:
    - Checks CLAUDE.md: where do features go?
    - Checks architecture: what exists for auth?
    - May load auth/SKILL.md if referenced
    - Begins planning

Step 3: Claude Code works
  Claude Code:
    - Creates files in correct locations (from CLAUDE.md)
    - Uses naming conventions (from CLAUDE.md)
    - Uses your error handling pattern (from CLAUDE.md)
    - Follows testing approach (from SKILL.md if loaded)
    - Runs tests via your slash command

Step 4: You review
  Files appear in VS Code editor.
  You review Claude Code changes.
  Approve, modify, or ask Claude Code to change approach.

Step 5: Iteration
  Claude Code runs tests, sees failures, iterates.
  You watch progress in VS Code and guide when needed.
```


### The Feedback Loop

**Why the feedback loop matters:**

```
Without CLAUDE.md:
  Claude Code writes code
  You check: Does it match conventions?   → Often no
  You explain what is wrong
  Claude Code fixes it
  Repeat until aligned

  You are reviewing TWO things:
    (1) Are the conventions right?
    (2) Is the logic right?

With CLAUDE.md:
  Claude Code writes code already matching conventions
  You check: Is the logic right?   → That is the only question

  You are reviewing ONE thing:
    (1) Is the logic right?

The feedback loop narrows from:
  "Is this right?" (conventions AND logic)
to:
  "Is the logic right?" (just logic)

This is the primary productivity gain from CLAUDE.md.
```

### Key Takeaways — Section 10

- Claude Code's **primary interface is the terminal (CLI)** — VS Code is integration layered on top
- You can work via the **terminal inside VS Code** or via the **VS Code extension** — both use the same tool
- Both methods read CLAUDE.md and Skill.md files in exactly the same way
- The core workflow is: session start → CLAUDE.md loaded → give task → review changes → iterate
- The key productivity benefit is **narrowing the feedback loop** from two questions (conventions + logic) to one (just logic)

## Section 11: Real-World Scenarios and Best Practices


### Scenario 1: Onboarding a New Developer

**The problem:**

```
New developer joins a team.
They need to learn:
  - How the project is structured
  - What conventions to follow
  - What tools to use
  - What is critical and what is flexible
  - How to work productively
```

**How CLAUDE.md changes onboarding:**

```
New developer sets up Claude Code and opens the project.
Claude Code reads CLAUDE.md.

New developer can ask:
  "Explain the project structure to me"
  → Claude Code reads the architecture section of CLAUDE.md
  → Explains the structure clearly and accurately

  "Show me how to create a new feature"
  → Claude Code reads conventions
  → Guides new developer through the correct project process

  "I want to add a test for this function"
  → Claude Code loads testing/SKILL.md (if referenced)
  → Shows the testing pattern for this specific project

  "What should I not touch without asking first?"
  → Claude Code reads constraints
  → Explains sensitive areas and who to consult

The CLAUDE.md becomes an interactive reference.
New developers get project-specific guidance,
not generic Claude guidance.
```

> **Accuracy note:** This is genuinely useful and does happen in practice. However,
> Claude Code can also make mistakes, and new developers should verify guidance against
> senior developer review. CLAUDE.md helps but does not replace mentorship.


### Scenario 2: Maintaining Consistency Across a Large Codebase

**The problem:**

```
Large codebase, multiple developers.
Each developer uses Claude Code.
Each developer explains context slightly differently.
Result: Inconsistent code, different patterns in different areas.
```

**How a shared CLAUDE.md helps:**

```
CLAUDE.md is committed to version control.
Every developer works from the same CLAUDE.md.
Claude Code reads the same conventions for everyone.

Developer A asks Claude Code to create a service.
Developer B asks Claude Code to create a service.
Both get the same pattern (from the shared CLAUDE.md).
Codebase stays consistent.

When conventions change:
  Team updates CLAUDE.md.
  All developers pull the change.
  Claude Code picks up new conventions immediately.
  No need to individually brief each person's Claude Code.

This is the team alignment benefit of CLAUDE.md.
```


### Scenario 3: Protecting Critical Code

**The problem:**

```
Critical payment processing code.
Must not be changed casually.
Must be peer-reviewed before any change.
But it is in a directory Claude Code can access.
```

**How constraints in CLAUDE.md help:**

```
CLAUDE.md states:
  "CONSTRAINT: The src/payments/ directory handles
   all payment processing. This code is subject to
   PCI DSS requirements.

   Do not modify any file in src/payments/ without:
   1. Explaining what you intend to change and why
   2. Receiving explicit confirmation to proceed

   Reason: Incorrect changes here can cause financial
   losses and compliance violations."

Result:
  When Claude Code considers touching payments/ code,
  it is instructed to pause and confirm.
  Claude Code reads the constraint and follows it.
```

> **Accuracy note:** This is effective instruction but not a hard lock.
> Always review Claude Code changes in critical areas.
> Version control is your true safety net.
> Never rely solely on CLAUDE.md constraints for security.

- Constraint sections in CLAUDE.md can **protect critical directories** by instructing Claude Code to pause and confirm

## Section 12: Common Mistakes and How to Avoid Them


### Mistake 1: Writing CLAUDE.md Once and Never Updating It

**The problem:**

```
Team writes CLAUDE.md when starting.
Six months later, conventions have evolved.
New patterns were adopted.
Some old patterns were abandoned.
But CLAUDE.md still reflects the old state.

Result:
  Claude Code gives outdated guidance.
  New developers learn old patterns from CLAUDE.md.
  Team trust in CLAUDE.md erodes.
```

**The solution:**

```
Treat CLAUDE.md like any other technical document:
  - Review it when conventions change
  - Update it when you adopt new patterns
  - Remove outdated sections
  - Assign ownership (who maintains it?)

A simple practice:
  Add CLAUDE.md review to your sprint or team retrospective.
  15 minutes every few weeks keeps it current.
```


### Mistake 2: Making CLAUDE.md Too Long

**The problem:**

```
Team adds everything to CLAUDE.md.
File becomes 500 lines, then 1000 lines.
Claude Code loads all of it every session.
Context window dominated by CLAUDE.md.
Limited room for actual code and conversation.
Information becomes hard to find.
```

**The solution:**

```
Keep core CLAUDE.md lean:
  - Architecture: essential overview only
  - Conventions: the ones that matter most
  - Constraints: the critical ones
  - Everything else: move to Skill.md files

Rough guideline:
  Core CLAUDE.md: under 200 lines for most projects

When CLAUDE.md approaches 200 lines, ask:
  "What here is only needed for specific tasks?"
  Move those sections to Skill.md files.
```


### Mistake 3: Writing Vague Conventions

**The problem:**

```
CLAUDE.md says: "Follow clean code principles"
                "Write readable functions"
                "Use meaningful names"

These are not actionable.
Claude Code already knows these principles.
It applies them according to its defaults,
not your team's specific interpretation.

Result: Code that is "good" but not matching YOUR style.
```

**The solution:**

```
Transform:
  "Use meaningful names"

Into:
  "Variable names should describe the data they hold.
   Prefer: userEmailAddress  over  email
   Prefer: isPaymentSuccessful  over  success
   Exception: loop counters can use i, j
   Exception: callback parameters often use err, result"
```

Every convention needs: a specific pattern, an example, and exceptions.


### Mistake 4: Listing Constraints Without Reasons

**The problem:**

```
CLAUDE.md says:
  - Never use array.forEach
  - Never use var
  - Never create files in src/legacy/

No reasons given.

Constraints without reasons feel arbitrary.
If the context makes them seem harmless,
Claude Code may not weight them heavily enough.
```

**The solution:**

```
Always include the reason:

  "Never use array.forEach
   Reason: We need to be able to break out of loops.
   Use for...of instead."

  "Never use var
   Reason: Block scoping prevents subtle bugs.
   Use const or let."

  "Never create files in src/legacy/
   Reason: This directory is being phased out.
   New code should not go here.
   Contact tech lead if you think it is needed."

Reasons help Claude Code weight constraints correctly.
Reasons help new developers understand the project.
Reasons document decisions for the future.
```


### Mistake 5: No Skill.md Files — Everything in One File

**The problem:**

```
All project knowledge in one CLAUDE.md.
File grows to 500+ lines.
All loaded every session.
Context window dominated by CLAUDE.md.
Limited room for actual work.
```

**The solution:**

```
When CLAUDE.md grows, ask:
  "Is this section only needed for specific tasks?"
  If yes → move to a Skill.md file.

Good candidates to move to Skill.md:
  - Testing (only needed when writing tests)
  - Deployment (only needed when deploying)
  - Database (only needed when writing queries or migrations)
  - Domain-specific logic (only needed for that domain)

Keep in main CLAUDE.md:
  - Architecture overview (everyone needs this)
  - Core conventions (needed for all code)
  - Critical constraints (always relevant)
  - Slash commands (used constantly)
  - Skills registry (pointer to Skill.md files)
```

---

### Key Takeaways — Section 12

- **Update CLAUDE.md regularly** — an outdated file is worse than no file, because it actively misleads
- **Keep CLAUDE.md under 200 lines** — move domain depth to Skill.md files to conserve context
- **Be explicit, not vague** — "follow clean code" is useless; "use camelCase, example: userName" is actionable
- **Always include reasons** for constraints — they help Claude Code weight them correctly and help humans understand
- **Create Skill.md files** when any CLAUDE.md section starts growing beyond its useful length

## Section 13: Summary and Next Steps


### What CLAUDE.md Actually Does (Accurate Summary)

```
CLAUDE.md provides written, persistent context to Claude Code.
This context is read at session start and applies throughout
the session.

The practical effect:
  - Less time explaining your project to Claude Code
  - More consistent code output aligned with your conventions
  - Clearer constraint awareness
  - Team alignment when everyone shares the same CLAUDE.md
  - Better onboarding experience for new developers

The honest limitations:
  - CLAUDE.md does not "program" Claude Code
    It provides context that Claude Code uses.

  - CLAUDE.md does not guarantee behavior
    It influences Claude Code responses.
    You should still review changes.

  - CLAUDE.md requires maintenance
    It becomes a liability if not kept current.

  - Context window is finite
    Very long CLAUDE.md files reduce available
    context for actual code and conversation.
```


### What Skills (Skill.md) Actually Do (Accurate Summary)

```
Skill.md files provide deep domain-specific context
that would make CLAUDE.md too long if included inline.

The practical effect:
  - Domain expertise available on demand
  - CLAUDE.md stays lean and focused
  - Detailed guidance for specific tasks
  - Easier to maintain (update one skill file without touching CLAUDE.md)

The honest limitations:
  - Skills must be explicitly loaded or triggered
    They do not auto-detect from task type
    unless your CLAUDE.md explicitly instructs this.

  - More files to maintain
    Each Skill.md needs updating when the domain changes.

  - Still occupies context window when loaded
    Loading many skills at once has a cost.
```

### What Slash Commands Actually Do (Accurate Summary)

```
Custom slash commands are prompt shortcuts defined in CLAUDE.md.

The practical effect:
  - Consistent instructions for repeated workflows
  - Faster to type than full instructions
  - Team-wide consistency when shared via version control
  - Good for quality gates (/code-review, /security-check)

The honest limitations:
  - Not shell scripts or macros
    They are prompt shortcuts. Claude Code acts on them as instructions.

  - Can be misinterpreted
    If the definition is ambiguous, results vary.

  - Need maintenance
    As workflows change, update your slash commands.
```
