# Understanding Embeddings

Before understanding ChromaDB, we must understand embeddings.

## What are Embeddings?

Embeddings are numerical vector representations of text.

AI models convert text into numbers while preserving meaning.

Example:

```text
"Dog" → [0.21, -0.45, 0.88, ...]
"Cat" → [0.19, -0.42, 0.85, ...]
"Car" → [-0.72, 0.11, -0.33, ...]
```

Notice:
- Dog and Cat vectors are close
- Car vector is far away

This allows AI systems to understand similarity.

# Introduction to FAISS

## What is FAISS?

FAISS stands for:

> Facebook AI Similarity Search

It is an open-source library developed by:

Facebook AI Research (FAIR)

FAISS is used for:
- fast vector similarity search
- nearest neighbor search
- semantic retrieval
- large-scale embedding search

It is one of the most widely used libraries for:
- Retrieval Augmented Generation (RAG)
- recommendation systems
- semantic search
- AI memory systems
- Large Language Model (LLM) applications

---

# Why Was FAISS Created?

Modern AI systems generate millions of embeddings.

Searching through all embeddings one-by-one becomes very slow.

Example:

Imagine:
- 10 million documents
- each converted into embeddings
- user asks a question

The system must quickly find:
> "Which embeddings are most similar?"

FAISS solves this problem efficiently.

---

# What Problem Does FAISS Solve?

Traditional databases are optimized for:
- exact matching
- SQL queries
- structured data

But AI applications need:
- similarity search
- nearest neighbor retrieval
- semantic matching

Example:

User asks:

> "How do AI models learn?"

FAISS should retrieve:
- neural networks
- backpropagation
- machine learning training
- gradient descent

even if exact words do not match.

---

# Understanding Similarity Search

FAISS works using:

> Vector Similarity Search

Instead of keyword matching.

---

# Example

Suppose we have embeddings:

```text
Dog → [0.11, 0.45, 0.88]
Cat → [0.10, 0.42, 0.86]
Car → [-0.72, 0.14, -0.31]
```

FAISS calculates:
- distance
- similarity
- nearest vectors

Result:
- Dog is close to Cat
- Dog is far from Car

This is called:
> Nearest Neighbor Search

---

# Core Concept of FAISS

FAISS stores vectors and allows us to:
- index vectors
- search nearest vectors
- retrieve similar embeddings

---

# High-Level Workflow

```text
Text
   ↓
Embedding Model
   ↓
Vectors
   ↓
FAISS Index
   ↓
Similarity Search
   ↓
Relevant Results
```

---

# What is an Index in FAISS?

An Index is the structure that stores vectors for efficient searching.

Think of it like:
- a search engine for embeddings

Example:

```python
index = faiss.IndexFlatL2(dimension)
```

The index stores vectors and performs similarity calculations.

---

# How FAISS Works

## Step 1 — Convert Text into Embeddings

Example:

```python
embedding = model.encode(text)
```

---

## Step 2 — Store Embeddings in FAISS

```python
index.add(vectors)
```

---

## Step 3 — Search Similar Embeddings

```python

    query_vector,
    k=3
)
```

FAISS returns:
- closest vectors
- similarity scores

# Important FAISS Concepts

# 1. Embeddings

Numerical representations of meaning.

Example:

```text
"AI" → [0.12, 0.88, ...]
```

---

# 2. Vectors

Embeddings stored as arrays.

Example:

```python
[0.21, 0.44, -0.67]
```

---

# 3. Similarity Search

Finding vectors closest in meaning.

---

# 4. Distance Metrics

FAISS calculates similarity using distance.

Common metrics:
- Euclidean Distance (L2)
- Cosine Similarity
- Inner Product

---

# Euclidean Distance (L2)

Measures geometric distance.

Example:

```python
faiss.IndexFlatL2(dimension)
```

---

# Inner Product Similarity

Used for cosine similarity style search.

Example:

```python
faiss.IndexFlatIP(dimension)
```

---

# Types of FAISS Indexes

FAISS provides multiple indexing methods.

---

# 1. IndexFlatL2

Simplest index.

Features:
- exact search
- accurate
- slower for huge datasets

Example:

```python
index = faiss.IndexFlatL2(dimension)
```

Best for:
- beginners
- small datasets
- learning

---

# 2. IndexFlatIP

Uses Inner Product similarity.

Useful for:
- cosine similarity
- embedding search

---

# 3. IVF (Inverted File Index)

Used for faster large-scale search.

Features:
- approximate nearest neighbor search
- scalable
- faster retrieval

Best for:
- enterprise AI systems

---

# 4. HNSW Index

Hierarchical graph-based search.

Features:
- very fast retrieval
- efficient memory usage

Used in:
- production AI systems

---

# Exact Search vs Approximate Search

| Exact Search | Approximate Search |
|---|---|
| More accurate | Faster |
| Slower | Less accurate |
| Small datasets | Massive datasets |

FAISS supports both.

````python
# ──────────────────────────────────────────────────────
# SECTION 2: Build a ChromaDB knowledge base (50 docs)
# ──────────────────────────────────────────────────────

import chromadb   # Vector database
import time       # For measuring query latency

# ── Step 1: Create client ──
# chromadb.Client() creates an in-memory instance (no files written to disk)
# For persistence use: chromadb.PersistentClient(path="./my_db")
client = chromadb.Client()

print("✅ ChromaDB client created (in-memory)")
print("-" * 60)

# ── Step 2: Define 50 sample documents ──
# These cover embeddings, vectors, ChromaDB, FAISS, search paradigms,
# the LLM Wiki pattern, and practical AI system design.

sample_documents = [
    # ── Embeddings (docs 0–9) ──
    "Embeddings are numerical representations of text that capture semantic meaning. "
    "Each word or sentence becomes a vector of numbers, where similar meanings "
    "produce vectors pointing in similar directions in high-dimensional space.",

    "All-MiniLM-L6-v2 produces 384-dimensional embeddings. "
    "All-mpnet-base-v2 produces 768-dimensional embeddings. "
    "Higher dimensions capture more nuance but require more computation.",

    "Cosine similarity measures the angle between two vectors. "
    "Smaller angles mean more similar meaning. "
    "A similarity score of 0.9 means the texts are nearly identical in meaning.",

    "During training, embedding models learn to place semantically similar texts "
    "close together in vector space. The model sees millions of sentence pairs "
    "and adjusts its weights to minimise the distance between similar pairs.",

    "Embeddings are context-aware: the word bank in river bank and bank account "
    "produces different vectors. Transformer models read surrounding context "
    "before encoding each word.",

    "An embedding is a fixed-length dense vector regardless of input length. "
    "A 1-word query and a 1,000-word document both produce a 384-float vector. "
    "This makes similarity comparison fast and uniform.",

    "Sentence embeddings encode an entire sentence as one vector. "
    "Word embeddings encode each word separately. "
    "For semantic search, sentence embeddings are usually more useful.",

    "Dimensionality reduction techniques like PCA or UMAP can project "
    "384-dimensional embeddings onto a 2D plane for visualisation. "
    "Clusters in the 2D view reveal groups of semantically similar documents.",

    "The embedding model is the foundation of every semantic search system. "
    "Choosing the right model — based on language, domain, and size — "
    "has a bigger impact on quality than the choice of vector database.",

    "Cross-encoder models re-rank top semantic search results by comparing "
    "query and document together. They are slower but more accurate than "
    "bi-encoder embedding models for the final ranking step.",

    # ── Vector Databases (docs 10–19) ──
    "ChromaDB simplifies semantic search by automatically handling embeddings. "
    "You add documents and ask questions in natural language. "
    "ChromaDB converts both to embeddings and finds similar documents.",

    "Vector databases like FAISS enable searching billions of vectors instantly. "
    "They use clever indexing techniques to avoid comparing with every vector. "
    "IndexIVFFlat clusters vectors; IndexIVFPQ compresses them further.",

    "Metadata filtering in ChromaDB lets you search with constraints. "
    "Filter by source, date, category, or any other metadata. "
    "This combines semantic similarity with structured search.",

    "ChromaDB includes vector search, full-text search, document storage, "
    "metadata filtering, and multi-modal retrieval in a single library. "
    "Its two-layer model uses a key/value store plus a vector index.",

    "FAISS was developed at Meta AI Research to solve internal search problems "
    "at Facebook scale. It can index 1.5 trillion vectors and is used in "
    "production across recommendation, search, and ads systems.",

    "Pinecone, Weaviate, Qdrant, and Milvus are managed vector databases "
    "for production deployments. They add features like replication, "
    "access control, and auto-scaling on top of core vector search.",

    "Approximate nearest-neighbour (ANN) algorithms sacrifice a small amount "
    "of accuracy for massive speed gains. FAISS IVFFlat is ANN. "
    "The trade-off is usually worth it: 0.5% accuracy loss, 1000x speed gain.",

    "ChromaDB automatically handles tokenization, embedding, and indexing. "
    "You can also supply your own embeddings from any model. "
    "Custom embeddings allow domain-specific models tuned for your data.",

    "A vector database is not just a search engine — it is a persistence layer "
    "for machine learning features. The same embeddings used for search "
    "can be used for recommendations, clustering, and anomaly detection.",

    "Hybrid search combines keyword (BM25) and vector (embedding) search. "
    "Some queries are better answered by exact keyword matches. "
    "Blending both scores with a weighted sum often outperforms either alone.",

    # ── Semantic Search (docs 20–29) ──
    "Semantic search differs from keyword search: it understands meaning. "
    "Keyword search finds exact word matches; semantic search finds related concepts. "
    "This is why semantic search works better for complex queries.",

    "RAG (Retrieval-Augmented Generation) uses embeddings to find relevant documents "
    "before generating answers. This grounds LLM responses in actual documents. "
    "The LLM reads retrieved documents and synthesises an answer.",

    "Vector databases are becoming essential for AI applications. "
    "They bridge the gap between unstructured text and structured search. "
    "Every semantic search engine, recommendation system, and RAG pipeline uses them.",

    "The retrieval step in RAG is critical: if the wrong documents are retrieved, "
    "the LLM generates an answer based on irrelevant context. "
    "Embedding model quality directly determines retrieval quality.",

    "Chunking strategy affects semantic search quality. "
    "Too-small chunks lose context; too-large chunks dilute the signal. "
    "Overlapping chunks (e.g. 256 tokens with 64-token overlap) work well.",

    "Re-ranking improves semantic search results. A fast bi-encoder retrieves "
    "the top 100 candidates; a slow cross-encoder re-ranks the top 10. "
    "This two-stage approach balances speed and accuracy.",

    "Dense retrieval (embedding-based) and sparse retrieval (BM25-based) "
    "capture different aspects of relevance. Dense retrieval understands synonyms; "
    "sparse retrieval handles rare keywords and exact product names.",

    "Evaluation metrics for semantic search include Precision@K, Recall@K, "
    "MRR (Mean Reciprocal Rank), and NDCG. Building a labelled evaluation set "
    "is the most important step before optimising a search system.",

    "Semantic search enables multilingual retrieval. A multilingual embedding model "
    "maps English and French sentences about the same topic to nearby vectors. "
    "Users can search in their native language and find documents in any language.",

    "Real-time indexing adds new documents to the vector database as they arrive. "
    "ChromaDB supports adding documents one by one or in batches. "
    "FAISS requires rebuilding or updating the index for new vectors.",

    # ── LLM Wiki Pattern (docs 30–39) ──
    "The LLM Wiki pattern is a way to build compounding knowledge. "
    "Instead of searching raw documents every time, an LLM maintains "
    "a structured wiki that grows more useful as you add sources.",

    "RAG retrieves. LLM Wiki compounds. Instead of repeatedly searching raw material, "
    "an LLM builds and maintains a persistent wiki that continuously integrates, "
    "organises, and refines knowledge over time.",

    "The LLM Wiki system rests on three directories: raw/ for source material "
    "(append-only), wiki/ for structured knowledge (LLM-maintained), "
    "and a schema file that governs conventions and workflows.",

    "An LLM Wiki page looks like a Wikipedia article in Markdown. "
    "It has a title, tags, a prose summary, related concept links, "
    "and citations back to the raw source files.",

    "The schema.md file in an LLM Wiki tells the LLM how to maintain the wiki: "
    "naming conventions, link syntax, how to handle contradictions, "
    "and when to create new pages vs update existing ones.",

    "LLM Wiki released as a GitHub Gist on April 4, 2026 and went viral. "
    "The idea of the idea file is that in the era of LLM agents, "
    "you share the idea, not specific code or an app.",

    "Most experience with LLMs and documents looks like RAG: you upload files, "
    "the LLM retrieves chunks at query time, and generates an answer. "
    "This works, but the LLM rediscovers knowledge from scratch on every question.",

    "LLM Wiki turns raw documents into a structured Markdown knowledge base "
    "that an LLM can query. It uses plain Markdown instead of vector databases, "
    "making it portable and editable by humans and LLMs alike.",

    "The LLM Wiki index.md file is a curated map of the entire wiki. "
    "It lists all pages with one-line summaries and groups them by topic. "
    "New users read index.md first to understand what knowledge exists.",

    "Linting in LLM Wiki checks for broken links, orphaned pages, "
    "outdated citations, and pages that have grown too long. "
    "The lint/ directory holds health-check scripts run periodically.",

    # ── AI System Design (docs 40–49) ──
    "Knowledge graphs and vector databases complement each other. "
    "Knowledge graphs capture structured relationships (A is a subtype of B). "
    "Vector databases capture unstructured semantic similarity.",

    "Token limits in LLMs make knowledge compression important. "
    "A 50-document corpus may contain 500,000 tokens — far more than fits in context. "
    "Chunking, embedding, and retrieval solve this scalability problem.",

    "Production AI search systems often combine: embedding model, vector database, "
    "keyword search, metadata filters, re-ranker, and LLM answer generation. "
    "Each layer improves quality at the cost of added latency.",

    "Fine-tuning an embedding model on domain-specific data improves retrieval "
    "significantly in specialised domains like medicine, law, and engineering. "
    "General models like all-MiniLM miss domain jargon and abbreviations.",

    "Caching frequent queries reduces latency in semantic search systems. "
    "If the same query (or a semantically identical one) was asked recently, "
    "return the cached result instead of re-embedding and re-searching.",

    "Monitoring search quality in production requires logging query embeddings, "
    "retrieved document IDs, user click-through rates, and explicit feedback. "
    "Without monitoring, search quality silently degrades over time.",

    "Versioning embedding models is critical. Switching from all-MiniLM-L6-v2 "
    "to all-mpnet-base-v2 changes all embeddings. The entire index must be rebuilt. "
    "Pin model versions in production code to avoid silent incompatibilities.",

    "Privacy-preserving search uses on-device embedding models so raw text "
    "never leaves the user device. Only the embedding vector is sent to the server. "
    "Apple, Google, and Meta use this approach in their products.",

    "Multimodal embeddings encode images, audio, and text in the same vector space. "
    "CLIP from OpenAI lets you search images with text queries. "
    "This powers reverse image search and cross-modal recommendations.",

    "The future of search combines real-time web retrieval, personal knowledge bases, "
    "and LLM reasoning. The LLM Wiki pattern is one approach to the personal layer. "
    "Vector databases provide the infrastructure; LLMs provide the intelligence.",
]

# Verify we have exactly 50 documents
assert len(sample_documents) == 50, f"Expected 50 docs, got {len(sample_documents)}"  # Hard check

print(f"✅ {len(sample_documents)} sample documents defined")
print(f"   Total characters: {sum(len(d) for d in sample_documents):,}")
print("-" * 60)

# ── Step 3: Create a ChromaDB collection and add documents ──
# A collection is like a table in a relational database
collection = client.create_collection(name="ai_knowledge_base")

# Build metadata for each document (category tags based on index)
def get_category(index):
    """
    Return a category string based on document index.

    Args:
        index (int): Document index 0–49

    Returns:
        str: Category label
    """
    if index < 10:       # Docs 0–9
        return "embeddings"
    elif index < 20:     # Docs 10–19
        return "vector_databases"
    elif index < 30:     # Docs 20–29
        return "semantic_search"
    elif index < 40:     # Docs 30–39
        return "llm_wiki"
    else:                # Docs 40–49
        return "ai_systems"

# Build lists required by ChromaDB
doc_ids       = [f"doc_{i}" for i in range(len(sample_documents))]  # ["doc_0", ..., "doc_49"]
doc_metadatas = [
    {
        "source": "knowledge_base",   # Which corpus this came from
        "category": get_category(i),  # Topic category
        "index": i                    # Original position
    }
    for i in range(len(sample_documents))
]

print("Adding 50 documents to ChromaDB...")
print("(ChromaDB auto-embeds each document using its default model)")

add_start = time.time()  # Start timing the add operation

# Add all documents in one batch call
collection.add(
    documents=sample_documents,    # Raw text — ChromaDB will embed these
    ids=doc_ids,                   # Unique string IDs
    metadatas=doc_metadatas        # Structured metadata for filtering
)

add_elapsed = time.time() - add_start  # Total time for adding + embedding

print(f"✅ {len(sample_documents)} documents added and embedded in {add_elapsed:.1f}s")
print(f"   Embeddings stored in vector index (ready for search)")
print("-" * 60)

# ── Step 4: Semantic search function ──
def semantic_search_chromadb(query, n_results=5, category_filter=None):
    """
    Search the ChromaDB collection using semantic similarity.

    Args:
        query           (str): Natural-language search query
        n_results       (int): How many results to return
        category_filter (str): Optional metadata filter (e.g. "embeddings")

    Returns:
        list[dict]: Ranked results with document text, similarity, and metadata
    """
    # Build optional where clause for metadata filtering
    where_clause = None                              # Default: no filter
    if category_filter is not None:                  # If a filter was supplied
        where_clause = {"category": category_filter} # Filter by category field

    # Run the query — ChromaDB embeds query automatically
    results = collection.query(
        query_texts=[query],         # ChromaDB converts this to an embedding
        n_results=n_results,         # Number of results requested
        where=where_clause           # Optional metadata filter
    )

    # Format results into readable dictionaries
    formatted_results = []

    for i in range(len(results["documents"][0])):  # Iterate over returned docs
        raw_distance    = results["distances"][0][i]       # L2 distance (lower = closer)
        similarity_score = 1.0 / (1.0 + raw_distance)     # Convert to 0–1 similarity

        result_dict = {
            "rank":       i + 1,                                         # 1-based rank
            "document":   results["documents"][0][i][:160] + "...",      # Truncated text
            "similarity": round(similarity_score, 3),                    # Similarity 0–1
            "distance":   round(raw_distance, 3),                        # Raw L2 distance
            "category":   results["metadatas"][0][i]["category"],        # Topic category
            "doc_id":     results["ids"][0][i],                          # Document ID
        }
        formatted_results.append(result_dict)  # Add to results list

    return formatted_results  # Return all formatted results


# ── Step 5: Keyword search function for comparison ──
def keyword_search(query, documents, n_results=5):
    """
    Simple keyword search — finds documents containing query words.

    Args:
        query     (str): Search query
        documents (list): List of document strings to search
        n_results (int): Max results to return

    Returns:
        list[dict]: Documents ranked by word-match count
    """
    query_words  = set(query.lower().split())  # Unique words in the query
    results_list = []                          # Accumulate matching documents

    for i, doc in enumerate(documents):          # Check every document
        doc_lower  = doc.lower()                  # Lowercase for matching
        match_count = 0                           # Matches found so far

        for word in query_words:                  # Test each query word
            if word in doc_lower:                 # Simple substring check
                match_count = match_count + 1     # Increment if found

        if match_count > 0:                       # Only include docs with ≥1 match
            results_list.append({
                "rank":          0,               # Filled in after sorting
                "document":      doc[:160] + "...",
                "matching_words": match_count,
                "score":         round(match_count / len(query_words), 3),
            })

    # Sort by number of matching words, highest first
    results_list.sort(key=lambda x: x["matching_words"], reverse=True)

    # Assign final ranks and trim to n_results
    for rank_index in range(len(results_list)):
        results_list[rank_index]["rank"] = rank_index + 1  # 1-based rank

    return results_list[:n_results]  # Return top-N only


# ── Step 6: Run demo searches and print comparison ──
demo_queries = [
    "How do I build a semantic search system?",
    "What is the LLM Wiki pattern?",
    "FAISS index types for large scale search",
]

for query in demo_queries:
    print(f"\n{'='*65}")
    print(f"QUERY: {query}")
    print(f"{'='*65}")

    # ── Keyword results ──
    print("\n🔤 KEYWORD SEARCH")
    print("-" * 65)
    kw_results = keyword_search(query, sample_documents)

    if len(kw_results) == 0:
        print("   ❌ No keyword matches found")
    else:
        for result in kw_results:
            print(f"   Rank {result['rank']} | score {result['score']:.2f} | {result['document'][:80]}...")

    # ── Semantic results ──
    print("\n🔍 SEMANTIC SEARCH (ChromaDB)")
    print("-" * 65)
    sem_results = semantic_search_chromadb(query, n_results=3)

    for result in sem_results:
        print(f"   Rank {result['rank']} | sim {result['similarity']:.3f} | [{result['category']}]")
        print(f"         {result['document'][:80]}...")

    print()

# ── Step 7: Demo metadata filtering ──
print("\n" + "="*65)
print("METADATA FILTER DEMO: Search only in category=embeddings")
print("="*65)

filtered_results = semantic_search_chromadb(
    query="how do vectors represent meaning",
    n_results=3,
    category_filter="embeddings"
)

for result in filtered_results:
    print(f"  Rank {result['rank']} | sim {result['similarity']:.3f} | {result['document'][:80]}...")

print("\n✅ Metadata filter restricted results to the embeddings category only")
````

# Building with FAISS ⚡

### Three Index Types Explained

| Index | Strategy | Speed | Accuracy | Memory | Best for |
|---|---|---|---|---|---|
| `IndexFlatL2` | Exhaustive (compare all) | Slow | 100% exact | Full | Tiny datasets |
| `IndexIVFFlat` | Cluster vectors, search nearest cluster | ~100× faster | ~99% | Full | Medium datasets |
| `IndexIVFPQ` | Cluster + compress (Product Quantisation) | ~1000× faster | ~95% | 10× less | Huge datasets |

FAISS uses **Voronoi cells** (k-means clusters) to partition the vector space.
At search time, only the nearest cluster(s) are scanned — not all vectors.
Product Quantisation (PQ) further compresses each vector to save memory.


````python
import faiss       # Facebook AI Similarity Search
import numpy as np # Numerical computing
import time        # Timing measurements

from sentence_transformers import SentenceTransformer  # Embedding model

print("Generating embeddings with all-MiniLM-L6-v2 for FAISS indexing...")
print("-" * 60)

# Re-use the model loaded in Section 2 (or reload if needed)
faiss_model = SentenceTransformer("all-MiniLM-L6-v2")  # 384-dimensional embeddings

# Embed all 50 sample documents
# Returns numpy array of shape (50, 384)
doc_embeddings = faiss_model.encode(sample_documents)

# FAISS requires float32 arrays
doc_embeddings_f32 = doc_embeddings.astype("float32")

print(f"✅ Embeddings generated: shape {doc_embeddings_f32.shape}")
print(f"   dtype : {doc_embeddings_f32.dtype}")
print(f"   size  : {doc_embeddings_f32.nbytes / 1024:.1f} KB")
print("-" * 60)

dimension  = doc_embeddings_f32.shape[1]  # 384 dimensions
n_docs     = doc_embeddings_f32.shape[0]  # 50 documents
n_clusters = 5                            # k-means clusters (keep small for 50 docs)

# ── Index 1: IndexFlatL2 — exhaustive exact search ──
print("\nBuilding IndexFlatL2 (exact search)...")
index_flat = faiss.IndexFlatL2(dimension)       # Exact L2 distance
index_flat.add(doc_embeddings_f32)               # Add all vectors
print(f"  ✅ IndexFlatL2 ready | vectors: {index_flat.ntotal}")

# ── Index 2: IndexIVFFlat — approximate with clustering ──
print("\nBuilding IndexIVFFlat (cluster-based approximate search)...")
quantiser      = faiss.IndexFlatL2(dimension)                                # Cluster centroids
index_ivf_flat = faiss.IndexIVFFlat(quantiser, dimension, n_clusters)        # IVF index
index_ivf_flat.train(doc_embeddings_f32)                                     # Learn clusters
index_ivf_flat.add(doc_embeddings_f32)                                       # Add vectors
print(f"  ✅ IndexIVFFlat ready | vectors: {index_ivf_flat.ntotal} | clusters: {n_clusters}")

# ── Index 3: IndexIVFPQ — approximate with compression ──
print("\nBuilding IndexIVFPQ (cluster + Product Quantisation)...")
# n_subquantizers must divide dimension (384): use 8 (384/8=48, valid)
n_subquantizers = 8   # Number of PQ sub-vectors
bits_per_code   = 5   # Bits per sub-vector code (changed from 8 to 5)
quantiser_pq    = faiss.IndexFlatL2(dimension)
index_ivf_pq    = faiss.IndexIVFPQ(
    quantiser_pq, dimension, n_clusters, n_subquantizers, bits_per_code
)
index_ivf_pq.train(doc_embeddings_f32)   # Learn clusters and codebook
index_ivf_pq.add(doc_embeddings_f32)     # Add compressed vectors
print(f"  ✅ IndexIVFPQ ready  | vectors: {index_ivf_pq.ntotal}")

# ── Comparison function ──
def compare_faiss_indexes(query_text, indexes_dict, k=3):
    """
    Run query on each FAISS index and print timing + top results.

    Args:
        query_text   (str): Natural-language query
        indexes_dict (dict): {name: faiss_index}
        k            (int): How many results to retrieve
    """
    print(f"\nQuery: \"{query_text}\"")
    print("-" * 65)
    print(f"{'Index':<18} {'Time (ms)':>10} {'Top dist':>10}  Top result snippet")
    print("-" * 65)

    # Embed the query
    query_vec = faiss_model.encode([query_text]).astype("float32")  # (1, 384)

    for index_name, index in indexes_dict.items():
        start    = time.perf_counter()                             # High-res timer start
        distances, indices = index.search(query_vec, k)            # Run search
        elapsed  = (time.perf_counter() - start) * 1000           # Convert to ms

        top_idx  = int(indices[0][0])                              # Best result index
        top_dist = float(distances[0][0])                          # Best result distance

        if top_idx == -1:                                          # -1 = not found (PQ edge case)
            snippet = "(no result)"
        else:
            snippet = sample_documents[top_idx][:50]               # First 50 chars

        print(f"{index_name:<18} {elapsed:>9.2f}  {top_dist:>10.3f}  {snippet}...")

    print("-" * 65)


# Group all indexes
faiss_indexes = {
    "IndexFlatL2":  index_flat,
    "IndexIVFFlat": index_ivf_flat,
    "IndexIVFPQ":   index_ivf_pq,
}

# Run comparison for two queries
compare_faiss_indexes("How do embeddings capture meaning?", faiss_indexes)
compare_faiss_indexes("FAISS indexing strategies for scale", faiss_indexes)

# ── Memory comparison ──
print("\nMEMORY COMPARISON")
print("-" * 65)

flat_mem  = n_docs * dimension * 4 / 1024         # 4 bytes per float32, in KB
ivf_mem   = flat_mem                               # IVFFlat stores full vectors
pq_mem    = n_docs * n_subquantizers * 1 / 1024   # 1 byte per code

print(f"{'IndexFlatL2':<18} {flat_mem:>8.2f} KB  (full float32 vectors)")
print(f"{'IndexIVFFlat':<18} {ivf_mem:>8.2f} KB  (full float32 vectors + centroids)")
print(f"{'IndexIVFPQ':<18} {pq_mem:>8.2f} KB  (compressed codes — {flat_mem/pq_mem:.0f}× smaller)")
print("-" * 65)
print("\nKey insight: IndexIVFPQ uses ~10-20× less memory with ~5% accuracy trade-off")
print("At 1 billion vectors, that difference is 768 GB vs ~50 GB!")
````
