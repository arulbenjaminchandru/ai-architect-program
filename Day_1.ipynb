## Section 1: What is an LLM?

An LLM (Large Language Model) is a system trained on large amounts of text to predict and generate language.

### Key Idea
LLMs do not "know" facts. They predict likely next tokens based on patterns.

### Real-world Uses
- Chat assistants
- Code generation
- Document summarization
- Search augmentation (RAG systems)

### Architecture View
LLM is:
- Stateless by default
- Context-driven
- Probabilistic system

## Diagram: LLM Interaction Flow
```
User Input → Tokenization → Model → Token Prediction Loop → Output
                          ↑
                     Context Window
```

## Section 2: Tokens

### What are Tokens?
Tokens are the units the model processes.

### Important Understanding
- Tokens ≠ words
- Tokens depend on tokenizer

### Example
"Artificial Intelligence" → ["Artificial", " Intelligence"]

### Why Tokens Matter (Exam Angle)
- Billing is token-based
- Context window is token-limited
- Prompt size affects performance

````python
# Basic token approximation (not real tokenizer)
text = "AI is transforming software engineering"
tokens = text.split()
print("Tokens:", tokens)
print("Token count:", len(tokens))
````

## Section 3: Training

### Key Concept
Model learns patterns from large text datasets.

### Training Phases
- Pretraining → general language understanding
- Fine-tuning → behavior shaping

### Important Clarification
- Model does NOT store documents
- Model learns statistical relationships

## Section 4: Prediction Mechanism

### Core Idea
LLM generates text one token at a time.

### Flow
1. Read input tokens
2. Predict next token
3. Append token
4. Repeat

### Why Responses Differ
- Probability-based selection
- Controlled by parameters

## Diagram: Token Generation Loop
```
Input → [Model]
        ↓
   Next Token
        ↓
 Append to Input
        ↓
 Repeat until stop
```

## Section 5: Key Parameters

### Temperature
- Low (0.0–0.3): deterministic
- Medium (0.5): balanced
- High (0.8–1.0): creative

### Context Window
- Max tokens model can process
- Includes prompt + response

### Max Tokens
- Limits output length

### Insight
- Deterministic use cases → low temperature
- Creative use cases → higher temperature

# 🧩 Big Picture

LLMs generate text **one token at a time**.

At each step:
👉 The model has multiple possible next tokens
👉 It assigns probabilities to each
👉 Parameters decide *how we pick from those probabilities*

## 🔑 Key Idea
Parameters = control knobs for behavior

# 🎲 Section 6: Temperature (The Creativity Knob)

## What it does
Controls **randomness** in token selection.

## Intuition
- LOW → safe, predictable
- HIGH → diverse, creative

## Mental Model
Imagine picking answers from a ranked list:
- Low temperature → always pick top answer
- High temperature → explore lower-ranked answers

# 🧠 Key Insight
- Temperature does NOT change knowledge
- It changes **how bold the model is in choosing outputs**

# 🎯 Section 7: Top-p (Nucleus Sampling)

## What it does
Limits token selection to a **probability mass**.

## Example
If top-p = 0.9:
👉 Only tokens contributing to top 90% probability are considered

## Intuition
- Temperature → randomness
- Top-p → selection boundary

# ⚖️ Temperature vs Top-p

| Concept     | Role |
|------------|------|
| Temperature | Controls randomness |
| Top-p       | Controls candidate pool |

## Practical Tip
👉 Usually tune ONE strongly, not both

# 📏 Section 8: Max Tokens

## What it does
Limits how long the output can be.

## Important
- Affects cost
- Affects completeness of answers

## Intuition
Think of it as a **word budget**

# 🧠 Section 9: Context Window

## What it is
Maximum tokens model can process at once.

Includes:
- Input prompt
- Chat history
- Output

## Key Insight
👉 When exceeded → older content is dropped

# 🧠 Section 10: Real-World Parameter Tuning

| Use Case          | Temperature | Top-p |
|-------------------|------------|-------|
| Chatbot           | 0.5        | 0.9   |
| Legal / Finance   | 0.1        | 0.8   |
| Creative Writing  | 0.9        | 1.0   |
| Summarization     | 0.3        | 0.9   |

# 🧾 Summary

- Temperature → randomness
- Top-p → selection boundary
- Max tokens → output length
- Context window → memory limit

## 🔑 Final Thought
LLMs are not controlled by prompts alone.
👉 Parameters define behavior.

## Final Summary

- LLM = next-token prediction system
- Tokens = core unit of processing
- Training = pattern learning
- Parameters = behavior control
- Architecture = system integration

### Certification Tip
Always think:
→ "How will this behave in production?"


# 🧾 1️⃣ FACTUAL QA (Low hallucination — banking use case)

### 🎯 Goal: Accurate, safe answer

**Prompt**

```
You are a banking assistant.

Answer only from the given context.
If the answer is not in the context, say "I don't know".

Context:
Savings account interest rate is 4% per annum.

Question:
What is the savings account interest rate?
```

**Parameters**

* Temperature: **0.1**
* Top-p: **0.9**
* Max tokens: **50**

**Why**

* Low temp → reduces randomness
* High top-p → still allows natural phrasing
* Small tokens → concise

**Expected Output**

```
The savings account interest rate is 4% per annum.
```

# 🧠 2️⃣ RAG (Context-grounded answer)

### 🎯 Goal: Strictly use retrieved data

**Prompt**

```
You are an AI assistant.

Rules:
- Only use the provided context
- Do not add external knowledge
- Answer in 2 sentences

Context:
The chatbot supports balance check, mini statement, and fund transfer.

Question:
What services are available?
```

**Parameters**

* Temperature: **0.2**
* Top-p: **0.8**
* Max tokens: **60**

**Why**

* Slight flexibility, but still controlled

**Expected Output**

```
The chatbot supports balance check, mini statement, and fund transfer.
These are the available services mentioned in the context.
```
